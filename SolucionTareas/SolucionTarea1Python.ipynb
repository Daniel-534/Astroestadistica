```python
import numpy as np
import matplotlib.pyplot as plt

# Diferentes tamaños de simulación
n_values = [10, 100, 1000, 10000, 100000]

# Crear subgráficos
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 15))
axes = axes.flatten()

for i, n in enumerate(n_values):
    
    # Simulación de dos dados
    dado1 = np.random.randint(1, 7, size=n)
    dado2 = np.random.randint(1, 7, size=n)
    sumas = dado1 + dado2
    
    # bins para datos discretos
    bins = np.arange(1.5, 13.5, 1)
    
    axes[i].hist(
        sumas,
        bins=bins,
        density=True,
        rwidth=0.8,
        color="skyblue",
        edgecolor="black"
    )
    
    axes[i].set_title(f"Simulación con n = {n}")
    axes[i].set_xlabel("Suma de los dados")
    axes[i].set_ylabel("Frecuencia relativa")
    axes[i].set_xticks(range(2,13))
    axes[i].grid(axis="y", linestyle="--", alpha=0.6)

# eliminar el espacio que sobra
fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```
