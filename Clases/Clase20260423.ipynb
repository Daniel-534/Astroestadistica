Una **prueba de hipótesis** es un método estadístico que utiliza datos de una muestra para decidir si una afirmación sobre un parámetro poblacional (o sobre la forma de una distribución) debe ser rechazada o no.

Los elementos fundamentales de este procedimiento son los siguientes:

### 1. Las Hipótesis Contradictorias
En toda prueba se comparan dos afirmaciones opuestas:
*   **Hipótesis Nula ($H_0$):** Es la pretensión que inicialmente se supone cierta o la "creencia previa". Siempre se formula como una igualdad respecto a un valor específico del parámetro (ej. $H_0: \mu = 100$).
*   **Hipótesis Alternativa ($H_a$):** Es la aseveración contradictoria a $H_0$ y suele representar la teoría que el investigador desea validar.

### 2. El Procedimiento de Decisión
Para decidir entre ambas, se definen dos componentes:
*   **Estadístico de prueba:** Una función de los datos muestrales (como $Z$ o $t$) que sirve como base para la decisión.
*   **Región de rechazo:** El conjunto de valores del estadístico de prueba que llevan a rechazar $H_0$ a favor de $H_a$.

### 3. Tipos de Error
Debido a que la decisión se basa en una muestra y no en toda la población, existen dos riesgos:
*   **Error de Tipo I ($\alpha$):** Consiste en rechazar la hipótesis nula cuando en realidad es verdadera. Al valor de $\alpha$ se le conoce como **nivel de significación**.
*   **Error de Tipo II ($\beta$):** Ocurre cuando no se rechaza la hipótesis nula a pesar de que es falsa.

### 4. El Valor P
Es el nivel de significación más pequeño al cual la hipótesis nula sería rechazada con los datos actuales. Un **valor P pequeño** (típicamente $\le 0.05$) indica que los datos son contradictorios a $H_0$, proporcionando evidencia para rechazarla.

Dentro de los procedimientos de prueba de hipótesis, el estadístico de prueba y el nivel de significancia son los componentes esenciales para la toma de decisiones.

### 1. Estadístico de Prueba
Es una función de los datos muestrales que sirve como base para decidir si se rechaza o no la hipótesis nula ($H_0$).
*   **Naturaleza:** Dado que se calcula a partir de una muestra, es una **variable aleatoria**; tiene una distribución de probabilidad específica cuando se asume que $H_0$ es verdadera (conocida como distribución nula).
*   **Función:** Suele medir la distancia entre el estimador puntual (obtenido de la muestra) y el valor propuesto en la hipótesis nula, a menudo expresada en unidades de desviación estándar.
*   **Ejemplos comunes:** Según el parámetro a probar, se utilizan estadísticos como $Z$ (para muestras grandes o varianza conocida), $t$ (para muestras pequeñas con varianza desconocida), $F$ (para comparar varianzas o en ANOVA) y $\chi^2$ (ji cuadrada para varianza o bondad de ajuste).

### 2. Nivel de Significancia ($\alpha$)
Es la probabilidad máxima tolerable de cometer un **error de tipo I**, es decir, rechazar la hipótesis nula cuando en realidad es verdadera.
*   **Control del error:** El investigador fija este valor de antemano para controlar el riesgo de una conclusión falsa a favor de la teoría que desea probar.
*   **Valores tradicionales:** Los niveles más utilizados son $0.10$, $0.05$ y $0.01$, dependiendo de la seriedad de las consecuencias de rechazar erróneamente $H_0$.
*   **Relación con la región de rechazo:** El nivel $\alpha$ determina los valores críticos que separan la región de rechazo de la de no rechazo; si el estadístico de prueba cae dentro de la región definida por $\alpha$, se rechaza $H_0$.

### Relación entre ambos
Un procedimiento de prueba se define seleccionando un **estadístico** y fijando un **nivel de significancia $\alpha$**. A menor nivel de significancia seleccionado, más grande debe ser el valor del estadístico de prueba para poder rechazar la hipótesis nula, lo que hace que la prueba sea más conservadora contra el rechazo accidental de la creencia previa.

La **región de rechazo** es el conjunto de todos los valores del estadístico de prueba para los cuales la hipótesis nula ($H_0$) es rechazada en favor de la hipótesis alternativa ($H_a$).

Aquí se detallan sus características fundamentales según el texto de Devore:

*   **Función en la toma de decisiones:** Un procedimiento de prueba de hipótesis se basa en un estadístico de prueba y una región de rechazo. Si el valor calculado del estadístico cae dentro de esta región, se decide rechazar $H_0$; de lo contrario, no se rechaza.
*   **Relación con el Nivel de Significancia ($\alpha$):** La región de rechazo se determina seleccionando un valor crítico que capture un área específica (el valor de $\alpha$) bajo la curva de distribución del estadístico cuando $H_0$ es verdadera. El nivel $\alpha$ representa la probabilidad de cometer un **error de tipo I** (rechazar $H_0$ siendo verdadera).
*   **Tipos de regiones según la hipótesis alternativa ($H_a$):**
    *   **Cola superior:** Se compone de valores grandes del estadístico. Se usa cuando $H_a$ indica que el parámetro es mayor que el valor nulo (ej. $H_a: \mu > \mu_0$).
    *   **Cola inferior:** Se compone de valores pequeños. Se usa cuando $H_a$ indica que el parámetro es menor que el valor nulo (ej. $H_a: \mu < \mu_0$).
    *   **Dos colas:** Se compone de valores muy alejados de la media en ambas direcciones. Se usa cuando $H_a$ indica que el parámetro es simplemente diferente al valor nulo (ej. $H_a: \mu \neq \mu_0$).
*   **Compromiso entre errores:** Al reducir el tamaño de la región de rechazo para disminuir la probabilidad de error tipo I ($\alpha$), inevitablemente aumenta la probabilidad de error tipo II ($\beta$, no rechazar $H_0$ cuando es falsa) para cualquier valor alternativo fijo.

Para realizar una prueba de hipótesis sobre la media poblacional ($\mu$), se deben seguir pasos analíticos que incluyen la formulación de hipótesis, la elección de un estadístico y la definición de una región de rechazo.

### Ejemplo 1: Prueba Z para $\mu$ ($\sigma$ conocida)
Este caso se aplica cuando la población es normal y la desviación estándar poblacional ($\sigma$) es un valor conocido.

1.  **Contexto:** Un fabricante afirma que la temperatura de activación promedio de un sistema de rociadores es de $130^{\circ}\text{F}$. Se toma una muestra de $n = 9$ con $\bar{x} = 131.08$ y se sabe que $\sigma = 1.5$.
2.  **Hipótesis:** Se prueba la hipótesis nula $H_0: \mu = 130$ frente a la alternativa $H_a: \mu \neq 130$.
3.  **Estadístico de prueba:** Se utiliza $z = \frac{\bar{x} - \mu_0}{\sigma/\sqrt{n}}$.
4.  **Cálculo:** $z = \frac{131.08 - 130}{1.5/\sqrt{9}} = \frac{1.08}{0.5} = 2.16$.
5.  **Región de rechazo:** Para un nivel de significancia $\alpha = 0.01$ en una prueba de dos colas, los valores críticos son $\pm z_{0.005} = \pm 2.58$.
6.  **Conclusión:** Como $2.16$ no cae en la región de rechazo (es decir, $-2.58 < 2.16 < 2.58$), no se rechaza $H_0$. No hay evidencia suficiente para contradecir la afirmación del fabricante.

### Ejemplo 2: Prueba $t$ para $\mu$ ($\sigma$ desconocida, muestra pequeña)
Este procedimiento es necesario cuando la población es normal pero $\sigma$ no se conoce, utilizando en su lugar la desviación estándar muestral ($s$).

1.  **Contexto:** Se estudia el peso máximo de izamiento para determinar si el promedio es mayor a $25\text{ kg}$. Se observa una muestra de $n = 5$ personas con $\bar{x} = 27.54$ y $s = 5.47$.
2.  **Hipótesis:** $H_0: \mu = 25$ frente a $H_a: \mu > 25$ (prueba de cola superior).
3.  **Estadístico de prueba:** Se utiliza $t = \frac{\bar{x} - \mu_0}{s/\sqrt{n}}$ con $n-1$ grados de libertad.
4.  **Cálculo:** $t = \frac{27.54 - 25}{5.47/\sqrt{5}} \approx 1.04$.
5.  **Región de rechazo:** Con $\alpha = 0.05$ y $\nu = 4$ grados de libertad, el valor crítico es $t_{0.05, 4} = 2.132$.
6.  **Conclusión:** Debido a que $1.04$ es menor que el valor crítico de $2.132$, no se rechaza $H_0$. No existe evidencia contundente de que el peso promedio de izamiento exceda los $25\text{ kg}$.

Para determinar si la magnitud de la estrella es diferente al valor de referencia ($m_0 = 5.58$), realizamos una **prueba de hipótesis para la media ($\mu$)** utilizando los pasos analíticos establecidos en el texto de Devore:

### 1. Formulación de Hipótesis
*   **Hipótesis nula ($H_0$):** $\mu = 5.58$ (La magnitud promedio verdadera es igual al valor de referencia).
*   **Hipótesis alternativa ($H_a$):** $\mu \neq 5.58$ (La magnitud promedio verdadera es diferente al valor de referencia; prueba de dos colas).

### 2. Datos y Estadísticos
*   **Valor nulo ($\mu_0$):** $5.58$.
*   **Desviación estándar ($\sigma$):** $1.39$.
*   **Tamaño de la muestra ($n$):** $100$ observaciones.
*   **Media muestral ($\bar{x}$):** $5.85$.

### 3. Cálculo del Estadístico de Prueba ($z$)
Dado que el tamaño de la muestra es grande ($n = 100 > 40$), el estadístico $Z$ tiene aproximadamente una distribución normal estándar:

$$z = \frac{\bar{x} - \mu_0}{\sigma/\sqrt{n}} = \frac{5.85 - 5.58}{1.39/\sqrt{100}}$$

1.  **Error estándar:** $1.39 / 10 = 0.139$.
2.  **Diferencia de medias:** $5.85 - 5.58 = 0.27$.
3.  **Valor de $z$:** $0.27 / 0.139 \approx \mathbf{1.94}$.

### 4. Análisis de la Región de Rechazo y Valor P
*   **Valores Críticos:** Para un nivel de significancia común de $\alpha = 0.05$ en una prueba de dos colas, los valores críticos son $\pm 1.96$.
*   **Comparación:** El valor calculado ($z = 1.94$) **no cae** en la región de rechazo, ya que está ligeramente dentro del intervalo $(-1.96, 1.96)$.
*   **Valor P:** El área en las colas para $z = 1.94$ es $2 \times [1 - \Phi(1.94)]$. Según la Tabla A.3, $\Phi(1.94) = 0.9738$, por lo que el valor P es aproximadamente **0.0524**.

### 5. Conclusión
Como el valor P ($0.0524$) es mayor que el nivel de significancia estándar $\alpha = 0.05$, **no se rechaza la hipótesis nula ($H_0$)**. 

**Respuesta al ejemplo:** No hay evidencia estadística suficiente (al nivel del 5%) para afirmar que la magnitud de la estrella es diferente a $5.58$. Aunque la media observada ($5.85$) es mayor, la diferencia puede atribuirse a la variabilidad natural del muestreo.
