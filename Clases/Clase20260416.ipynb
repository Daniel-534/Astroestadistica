Si $n$ no es suficientemente grangre y $X\sim \text{Normal}$

Si $n$ no es suficientemente grande (típicamente $n < 40$ en este contexto) y la población de la cual se extrae la muestra sigue una distribución **Normal**, el Teorema del Límite Central no puede ser invocado para asegurar que la media muestral siga una distribución normal si se desconoce la desviación estándar poblacional ($\sigma$).

En este escenario, para realizar inferencias sobre la media poblacional $\mu$, las fuentes establecen el uso de la **distribución $t$ de Student** en lugar de la distribución normal estándar ($Z$).

### 1. El estadístico $T$
Cuando la población es normal, la variable estandarizada definida como:
$$T = \frac{\bar{X} - \mu}{S/\sqrt{n}}$$
sigue una distribución de probabilidad denominada **distribución $t$ con $n - 1$ grados de libertad (gl)**.

### 2. Justificación analítica
*   **Incertidumbre en la variabilidad:** Cuando el tamaño de la muestra $n$ es pequeño, la desviación estándar muestral ($S$) no es una aproximación precisa de $\sigma$. 
*   **Dispersión de la curva:** Analíticamente, la distribución $t$ está más "esparcida" que la curva normal estándar ($Z$) para reflejar la variabilidad extra que surge de tener que estimar tanto $\mu$ como $\sigma$ a partir de una muestra pequeña.
*   **Grados de libertad ($\nu$):** El parámetro gl se define como $\nu = n - 1$ porque, aunque el estadístico se basa en $n$ desviaciones, la suma de estas desviaciones con respecto a la media es cero ($\sum(X_i - \bar{X}) = 0$), lo que significa que solo $n-1$ de ellas se determinan libremente.

### 3. Aplicación al Intervalo de Confianza
Para calcular un **intervalo de confianza del $100(1-\alpha)\%$** bajo estas condiciones, se utiliza la expresión:
$$\bar{x} \pm t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}}$$.

Donde $t_{\alpha/2, n-1}$ es el valor crítico obtenido de la **Tabla A.5** (o Tabla 7.3), que captura un área de $\alpha/2$ en la cola superior de la curva $t$ con $n-1$ grados de libertad.

### 4. Robustez del método
Es importante notar que este intervalo basado en la distribución $t$ es **robusto** ante alejamientos pequeños o moderados de la normalidad en la población original, siempre que la muestra no sea extremadamente pequeña y la distribución no sea altamente asimétrica. Sin embargo, si existen dudas serias sobre la normalidad de la población, el texto sugiere considerar técnicas alternativas como el *bootstrap* o procedimientos no paramétricos.

El análisis de los intervalos de confianza para la media poblacional ($\mu$) varía significativamente dependiendo del tamaño de la muestra ($n$) y de si se conoce o no la variabilidad de la población ($\sigma$). A continuación, se detallan los dos casos solicitados basándose en los principios de inferencia estadística del texto de Devore.

### 1. Caso en el que $n$ es pequeño, pero $\sigma$ es conocido

Aunque el autor señala que este escenario es "un tanto irreal" porque en la práctica es difícil conocer la dispersión ($\sigma$) sin conocer el centro ($\mu$), su desarrollo analítico es fundamental para la teoría.

*   **Suposición de Normalidad:** Para que este método sea válido con una muestra pequeña, es indispensable suponer que la **distribución de la población es normal**. Bajo esta premisa, independientemente del tamaño de la muestra ($n$), la media muestral ($\bar{X}$) seguirá una distribución normal.
*   **Estadístico de Prueba:** Se utiliza la variable normal estándar $Z$, ya que al conocer $\sigma$, el denominador de la estandarización es una constante fija:
    $$Z = \frac{\bar{X} - \mu}{\sigma/\sqrt{n}} \text{}$$
*   **Construcción del Intervalo:** Al ser la distribución de $Z$ perfectamente conocida, se utilizan los valores críticos de la tabla normal estándar ($z_{\alpha/2}$). El intervalo resultante es:
    $$IC = \bar{x} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \text{}$$
*   **Propiedad:** La precisión del intervalo está determinada por el valor conocido de $\sigma$ y no se ve afectada por la incertidumbre de estimar la varianza desde la muestra.

### 2. Caso de $n$ grande y $\sigma$ desconocido

Este es el escenario más común en la práctica estadística, donde el investigador no conoce la desviación estándar poblacional y debe recurrir a la información de la muestra.

*   **Teorema del Límite Central (TLC):** Cuando el tamaño de muestra es grande, el TLC garantiza que $\bar{X}$ tiene aproximadamente una distribución normal, sin importar la forma de la distribución de la población original.
*   **Sustitución de $\sigma$ por $s$:** Al no conocer $\sigma$, se utiliza la desviación estándar muestral ($s$) como estimador puntual. El autor indica que en muestras grandes, la variabilidad extra introducida por usar $s$ es mínima, por lo que la variable estandarizada sigue aproximándose a una normal estándar ($Z$).
*   **Regla del Tamaño de Muestra:** Para justificar el uso de este intervalo con $\sigma$ desconocido, el texto sugiere una regla empírica más conservadora que la del TLC usual, recomendando que **$n \ge 40$**.
*   **Fórmula del Intervalo:** Se emplea el valor crítico $z_{\alpha/2}$ en combinación con el error estándar estimado ($s/\sqrt{n}$):
    $$IC = \bar{x} \pm z_{\alpha/2} \cdot \frac{s}{\sqrt{n}} \text{}$$

El siguiente cuadro ilustra los métodos empleados en cada situación para la estimación de intervalos de confianza para la media $\mu$:

Si se tiene una población con distribución **Normal** ($X_i \sim \text{Normal}$), un tamaño de muestra **$n$ pequeño** (generalmente $n < 40$) y la desviación estándar poblacional **$\sigma$ es desconocida**, el procedimiento estadístico estándar consiste en utilizar la **distribución $t$ de Student** en lugar de la distribución normal estándar ($Z$).

Bajo estas condiciones, las fuentes detallan lo siguiente:

### 1. El Estadístico $T$
Cuando $\sigma$ es desconocido, se debe estimar mediante la desviación estándar muestral $S$. Al no ser $n$ lo suficientemente grande para que la variabilidad de $S$ sea insignificante, la variable estandarizada resultante no sigue una distribución normal, sino una distribución $t$:
$$T = \frac{\bar{X} - \mu}{S/\sqrt{n}}$$
Este estadístico tiene una distribución de probabilidad con **$n - 1$ grados de libertad** (denotados como $\nu$ o $gl$).

### 2. Justificación Analítica
*   **Variabilidad extra:** En muestras pequeñas, $S$ puede no aproximarse bien a $\sigma$. La distribución $t$ está más "esparcida" que la curva $Z$ para reflejar esta incertidumbre adicional proveniente del denominador.
*   **Grados de libertad:** Se utiliza $n-1$ porque la suma de las desviaciones con respecto a la media muestral es siempre cero, lo que implica que solo $n-1$ de esas desviaciones se determinan libremente.
*   **Convergencia:** A medida que el número de grados de libertad aumenta, la dispersión de la distribución $t$ disminuye y la curva comienza a parecerse a la normal estándar ($Z$).

### 3. Aplicaciones Prácticas

#### Intervalo de Confianza (IC)
Para estimar la media $\mu$ con un nivel de confianza del $100(1-\alpha)\%$, se utiliza la fórmula:
$$IC = \bar{x} \pm t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}}$$
Donde $t_{\alpha/2, n-1}$ es el valor crítico que captura el área central deseada en la tabla de la distribución $t$.

#### Pruebas de Hipótesis
Para probar una hipótesis nula $H_0: \mu = \mu_0$, el valor del estadístico de prueba se calcula como:
$$t = \frac{\bar{x} - \mu_0}{s/\sqrt{n}}$$
La región de rechazo se basa en los valores críticos $t_{\alpha, n-1}$ (para pruebas unilaterales) o $t_{\alpha/2, n-1}$ (para pruebas bilaterales) obtenidos de la tabla correspondiente.

### Resumen de condiciones para el Caso III (Devore)
Para que este método sea válido cuando $n$ es pequeño y $\sigma$ es desconocido, es indispensable que la población de interés sea **normal**. Si la población fuera considerablemente no normal y $n$ pequeño, se tendrían que utilizar métodos alternativos como procedimientos "libres de distribución" o técnicas de *bootstrap*.

| :--- | :--- | :--- | :--- | :--- | :--- |
| **I** | Pequeño o grande | **Conocida** | Debe ser Normal (si $n$ es pequeño) | Normal Estándar ($Z$) | $\bar{x} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$ |
| **II** | Grande ($n \ge 40$) | **Desconocida** (se usa $s$) | Cualquiera (gracias al TLC) | Normal Estándar ($Z$ aprox.) | $\bar{x} \pm z_{\alpha/2} \cdot \frac{s}{\sqrt{n}}$ |

### Detalles Analíticos Clave:

1.  **Caso I (Ideal pero poco frecuente):** Se utiliza cuando se conoce la variabilidad histórica o teórica de la población ($\sigma$). Si la muestra es pequeña, el método solo es válido si la población de origen es normal.
2.  **Caso II (Muestra Grande):** Es el más común en la práctica. Al ser $n \ge 40$, el error introducido al estimar $\sigma$ con la desviación estándar muestral ($s$) es lo suficientemente pequeño como para seguir usando los valores críticos de la tabla normal estándar ($z_{\alpha/2}$).
3.  **Caso III (Población Normal con muestra pequeña):** Cuando no se puede invocar el Teorema del Límite Central (muestra pequeña) y no se conoce $\sigma$, se debe compensar la incertidumbre extra utilizando la **distribución $t$**. Esta curva está más "esparcida" que la normal para reflejar que $s$ puede no ser una estimación precisa de $\sigma$ en muestras cortas. Es vital verificar que la población sea normal (mediante una gráfica de probabilidad normal) para que este intervalo sea válido.

**Nota sobre los Grados de Libertad ($\nu$):** En el Caso III, se usa $n-1$ porque, aunque el cálculo de $s$ se basa en $n$ desviaciones, solo $n-1$ de ellas pueden variar libremente (la suma de todas las desviaciones respecto a la media es siempre cero).

### 1. Estimador Insesgado ($\hat{\theta} \sim$ Insesgado)
La primera condición exige que el estadístico seleccionado ($\hat{\theta}$) sea un estimador insesgado del parámetro $\theta$.
*   **Explicación:** Esto significa que el valor esperado del estimador es igual al parámetro verdadero ($E(\hat{\theta}) = \theta$). Analíticamente, esto garantiza que la distribución de muestreo del estimador esté siempre **centrada** en el valor real que se desea capturar.

### 2. Distribución Independiente de Parámetros Desconocidos
La derivación de un IC requiere encontrar una variable aleatoria funcional (como $Z$ o $T$) que dependa de la muestra y de $\theta$, pero cuya **distribución de probabilidad sea conocida** y no dependa de parámetros desconocidos.
*   **Explicación:** Si $\hat{\theta}$ tiene una distribución aproximadamente normal (lo cual suele ocurrir en muestras grandes por el TLC), la variable estandarizada $Z = (\hat{\theta} - \theta)/\sigma_{\hat{\theta}}$ sigue una distribución normal estándar $N(0,1)$, la cual es independiente de $\theta$. Esto permite usar valores críticos fijos de tablas (como $z_{\alpha/2}$) para definir los límites.

### 3. Error Estándar ($\sigma_{\hat{\theta}}$ o $\hat{\sigma}_{\hat{\theta}}$)
Para cuantificar la precisión, se requiere la desviación estándar del estimador, denominada **error estándar**.
*   **Caso $\sigma_{\hat{\theta}}$ (Conocido):** Si la variabilidad es conocida, se usa directamente en la fórmula.
*   **Caso $\hat{\sigma}_{\hat{\theta}}$ o $s_{\hat{\theta}}$ (Estimado):** Si el error estándar implica parámetros desconocidos, se sustituyen por sus estimaciones muestrales (por ejemplo, usar $s$ en lugar de $\sigma$). En muestras grandes ($n \ge 40$), esta sustitución mantiene la validez de la distribución normal.

---

### Conclusión: La Expresión General
Al combinar estos tres elementos, el intervalo de confianza de $100(1-\alpha)\%$ para $\theta$ se expresa de forma compacta como:

$$\hat{\theta} \pm z_{\alpha/2} \cdot \sigma_{\hat{\theta}} \quad \text{o} \quad \hat{\theta} \pm z_{\alpha/2} \cdot \hat{\sigma}_{\hat{\theta}}$$

Donde:
*   **$\hat{\theta}$**: Es el centro del intervalo (estimación puntual).
*   **$z_{\alpha/2}$**: Es el valor crítico que garantiza la fiabilidad deseada.
*   **$z_{\alpha/2} \cdot \sigma_{\hat{\theta}}$** Es el **margen de error** o límite en el error de estimación.

# Ejemplo 1

Para ilustrar la estimación de una proporción poblacional bajo el modelo binomial, utilizaremos un caso basado en el control de calidad de componentes semiconductores, adaptando los datos del **Ejercicio 19** de la Sección 7.2 de las fuentes.

### 1. Contexto del Problema: Inspección de Troqueles de Obleas
En la manufactura de semiconductores, es vital conocer la proporción de unidades que funcionan correctamente después de ser grabadas en una oblea. Se realiza un estudio de un proceso de inspección particular donde se examina una muestra de troqueles mediante una sonda de prueba.

*   **Población:** Todos los troqueles producidos bajo un proceso estable.
*   **Variable Aleatoria ($X$):** El número de troqueles que pasan la prueba de la sonda. En este modelo, cada troquel es un ensayo independiente con dos resultados posibles: pasar ("éxito") o fallar ("fracaso").
*   **Distribución:** Dado que el tamaño de la población es mucho mayor que la muestra, se asume que $X$ sigue una distribución binomial: $X \sim \text{bin}(n; p)$, donde $n$ es el número de ensayos y $p$ es la proporción verdadera de troqueles que pasan la prueba.

### 2. Datos Obtenidos de la Muestra
Según el registro del estudio:
*   **Tamaño de la muestra ($n$):** $356$ troqueles examinados.
*   **Número de éxitos ($X$):** $201$ troqueles pasaron la prueba.

### 3. Estimación Puntual ($\hat{p}$)
El estimador de la proporción poblacional $p$ es la proporción muestral de éxitos. Analíticamente se define como:
$$\hat{p} = \frac{X}{n}$$
Sustituyendo los valores observados:
$$\hat{p} = \frac{201}{356} \approx \mathbf{0.5646}$$
Esto significa que, puntualmente, estimamos que el **56.46%** de todos los troqueles en la población pasarán la prueba.

### 4. Construcción del Intervalo de Confianza (IC)
Para cuantificar la fiabilidad de esta estimación, calcularemos un **intervalo de confianza del 95%**.

#### A. Verificación de condiciones para muestra grande
Las fuentes indican que este procedimiento es válido si tanto $n\hat{p} \ge 10$ como $n(1-\hat{p}) \ge 10$.
*   $n\hat{p} = 201 \ge 10$ (Cumple).
*   $n\hat{q} = 356 - 201 = 155 \ge 10$ (Cumple).
Dado que $n = 356$ excede considerablemente el umbral de $n \ge 40$, la distribución de $\hat{p}$ es aproximadamente normal.

#### B. Selección del valor crítico ($z$)
Para un nivel de confianza del 95%, el valor crítico que captura el área central de la curva normal estándar es **$z_{0.025} = 1.96$**.

#### C. Cálculo del Error Estándar Estimado
La desviación estándar estimada del estimador $\hat{p}$ (error estándar) se calcula como:
$$\sigma_{\hat{p}} \approx \sqrt{\frac{\hat{p}\hat{q}}{n}} = \sqrt{\frac{0.5646 \cdot (1 - 0.5646)}{356}}$$
$$\sigma_{\hat{p}} \approx \sqrt{\frac{0.2458}{356}} \approx \mathbf{0.0263}$$

#### D. Límites del Intervalo Tradicional
Utilizando la fórmula general $\hat{p} \pm z_{\alpha/2} \cdot \sigma_{\hat{p}}$:
*   **Límite Inferior:** $0.5646 - (1.96 \cdot 0.0263) \approx 0.5131$
*   **Límite Superior:** $0.5646 + (1.96 \cdot 0.0263) \approx 0.6161$

### 5. Interpretación del Resultado
Con un **nivel de confianza del 95%**, se estima que la proporción verdadera de troqueles que pasan la prueba ($p$) se encuentra entre **0.5131 y 0.6161**. 

En términos prácticos, el fabricante puede afirmar con un 95% de seguridad que la tasa de éxito de su proceso de producción de chips no es inferior al 51.3% ni superior al 61.6%. Si este procedimiento de muestreo se repitiera indefinidamente, el 95% de los intervalos calculados de esta forma contendrían la proporción real de la población.

$$P(z_{\alpha/2}\leq \frac{\hat{p}-p}{\sqrt{p(1-p)/n}}\leq z_{1-\alpha/2}) = 1-\alpha$$

Esta expresión es la base fundamental para construir el **intervalo de confianza para una proporción poblacional ($p$)**. Proviene de la estandarización de la distribución muestral de la proporción.

### 1. ¿De dónde sale la expresión? (Justificación Teórica)

La lógica detrás de esta probabilidad se basa en tres puntos clave detallados en el libro de Devore:

*   **Normalidad de $\hat{p}$:** El estimador puntual de la proporción poblacional es $\hat{p} = X/n$. Cuando el tamaño de muestra $n$ es suficientemente grande (específicamente cuando $np \geq 10$ y $n(1-p) \geq 10$), la variable aleatoria $\hat{p}$ sigue una **distribución aproximadamente normal**.
*   **Parámetros de la distribución:** El valor esperado de $\hat{p}$ es la proporción real $p$ ($E(\hat{p}) = p$) y su desviación estándar (error estándar) es $\sigma_{\hat{p}} = \sqrt{p(1-p)/n}$.
*   **Estandarización:** Al restar su media ($p$) y dividir por su error estándar, obtenemos una variable $Z$ que se distribuye como una **normal estándar ($N(0,1)$)**:
    $$Z \approx \frac{\hat{p}-p}{\sqrt{p(1-p)/n}}$$
*   **Área de Probabilidad:** Para un nivel de confianza $1-\alpha$, buscamos los valores críticos $-z_{\alpha/2}$ y $z_{\alpha/2}$ (en tu expresión anotado como $z_{1-\alpha/2}$) que capturen el área central de la curva normal.

---

### 2. Despeje de $p$ paso a paso

A diferencia de la media $\mu$, despejar $p$ es más complejo porque el parámetro aparece tanto en el numerador como en el denominador (dentro de la raíz). El texto de Devore explica que para obtener los límites de confianza, debemos resolver la desigualdad resultante para $p$.

Para simplificar, usaremos $z$ en lugar de $z_{\alpha/2}$. Partimos de la igualdad en los límites:
$$\frac{|\hat{p} - p|}{\sqrt{p(1-p)/n}} = z$$

#### Paso 1: Elevar al cuadrado para eliminar la raíz
$$ \frac{(\hat{p} - p)^2}{p(1-p)/n} = z^2 $$

#### Paso 2: Reorganizar la ecuación
Multiplicamos ambos lados por el denominador:
$$ (\hat{p} - p)^2 = \frac{z^2}{n} \cdot p(1-p) $$

#### Paso 3: Expandir los términos
A la izquierda expandimos el binomio y a la derecha distribuimos:
$$ \hat{p}^2 - 2\hat{p}p + p^2 = \frac{z^2}{n}p - \frac{z^2}{n}p^2 $$

#### Paso 4: Agrupar términos para formar una ecuación cuadrática ($Ap^2 + Bp + C = 0$)
Llevamos todos los términos con $p$ a un solo lado:
$$ p^2 + \frac{z^2}{n}p^2 - 2\hat{p}p - \frac{z^2}{n}p + \hat{p}^2 = 0 $$
Factorizamos $p^2$ y $p$:
$$ \left( 1 + \frac{z^2}{n} \right)p^2 - \left( 2\hat{p} + \frac{z^2}{n} \right)p + \hat{p}^2 = 0 $$

#### Paso 5: Aplicar la fórmula cuadrática
Identificamos los coeficientes:
*   $A = 1 + \frac{z^2}{n}$
*   $B = -\left( 2\hat{p} + \frac{z^2}{n} \right)$
*   $C = \hat{p}^2$

Usamos la fórmula $p = \frac{-B \pm \sqrt{B^2 - 4AC}}{2A}$. Tras realizar las manipulaciones algebraicas y simplificaciones de los términos bajo la raíz, se llega a la expresión de los **límites de confianza** que presenta el libro:

#### Paso 6: Resultado final (Límites de Wilson / Score)
$$p = \frac{\hat{p} + \frac{z^2}{2n} \pm z\sqrt{\frac{\hat{p}\hat{q}}{n} + \frac{z^2}{4n^2}}}{1 + z^2/n}$$

Este intervalo es conocido como el **intervalo de Wilson** o de puntaje (score), y es más preciso que el intervalo tradicional, especialmente cuando $n$ no es muy grande o $p$ está cerca de 0 o 1. El signo **$-$** da el límite inferior y el **$+$** el límite superior.

La construcción de un intervalo de confianza para la varianza poblacional ($\sigma^2$) se fundamenta en el análisis de la distribución de la varianza muestral ($S^2$) cuando se extraen muestras de una población con distribución normal. A diferencia de los intervalos para la media, este procedimiento es altamente sensible a la suposición de normalidad de la población original.

### 1. Suposiciones fundamentales
Para que el desarrollo teórico sea válido, se debe partir de las siguientes premisas:
*   Se dispone de una muestra aleatoria $X_1, X_2, \dots, X_n$ de tamaño $n$.
*   La población de origen sigue una distribución normal con media $\mu$ y varianza $\sigma^2$.
*   El parámetro de interés es $\sigma^2$, mientras que $\mu$ suele ser desconocido.

### 2. El estadístico de prueba y su distribución
El fundamento analítico para este intervalo es el estadístico definido por la relación entre la varianza muestral y la poblacional:
$$\frac{(n-1)S^2}{\sigma^2}$$
Según el teorema fundamental presentado en las fuentes, esta variable aleatoria sigue una **distribución ji cuadrada ($\chi^2$)** con $\nu = n-1$ grados de libertad. A diferencia de la distribución normal o la $t$ de Student, la curva ji cuadrada no es simétrica y solo asume valores positivos, lo que requiere una notación específica para sus valores críticos.

### 3. Selección de valores críticos
Para un nivel de confianza deseado de $100(1-\alpha)\%$, se deben identificar dos puntos en el eje horizontal de la distribución ji cuadrada que capturen el área central:
*   **$\chi^2_{\alpha/2, n-1}$:** El valor crítico que deja un área de $\alpha/2$ en la cola superior (derecha).
*   **$\chi^2_{1-\alpha/2, n-1}$:** El valor crítico que deja un área de $1-\alpha/2$ a su derecha, lo que equivale a dejar $\alpha/2$ en la cola inferior (izquierda).

### 4. Derivación analítica del intervalo
El desarrollo matemático comienza con la siguiente proposición de probabilidad:
$$P\left( \chi^2_{1-\alpha/2, n-1} \le \frac{(n-1)S^2}{\sigma^2} \le \chi^2_{\alpha/2, n-1} \right) = 1 - \alpha$$.

Para aislar el parámetro $\sigma^2$ en el centro de la desigualdad, se realizan las siguientes manipulaciones algebraicas:
1.  Se toma el recíproco de cada término, lo que invierte el sentido de las desigualdades.
2.  Se multiplica toda la expresión por la cantidad constante $(n-1)S^2$.
3.  Al reordenar los términos para mantener el formato de límite inferior a la izquierda y límite superior a la derecha, se obtiene la expresión final.

### 5. Definición formal del intervalo
El intervalo de confianza del $100(1-\alpha)\%$ para la varianza $\sigma^2$ queda definido por los siguientes límites:
*   **Límite Inferior de Confianza (LIC):** $\frac{(n-1)s^2}{\chi^2_{\alpha/2, n-1}}$.
*   **Límite Superior de Confianza (LSC):** $\frac{(n-1)s^2}{\chi^2_{1-\alpha/2, n-1}}$.

### 6. Interpretación conceptual
El resultado de este procedimiento no es un valor exacto para $\sigma^2$, sino un rango de valores factibles. La interpretación técnica dicta que si este experimento de muestreo se repitiera una cantidad infinita de veces, el $100(1-\alpha)\%$ de los intervalos calculados mediante esta fórmula lograrían capturar el verdadero valor de la varianza poblacional. Si el intervalo resultante es estrecho, indica que la variabilidad de la población ha sido estimada con alta precisión; si es ancho, refleja una gran incertidumbre debida habitualmente a un tamaño de muestra pequeño o a una variabilidad sustancial en los datos observados.

La construcción de un intervalo de confianza para la varianza poblacional ($\sigma^2$) se fundamenta en el análisis de la distribución de la varianza muestral ($S^2$) cuando se extraen muestras de una población con distribución normal. A diferencia de los intervalos para la media, este procedimiento es altamente sensible a la suposición de normalidad de la población original.

### 1. Suposiciones fundamentales
Para que el desarrollo teórico sea válido, se debe partir de las siguientes premisas:
*   Se dispone de una muestra aleatoria $X_1, X_2, \dots, X_n$ de tamaño $n$.
*   La población de origen sigue una distribución normal con media $\mu$ y varianza $\sigma^2$.
*   El parámetro de interés es $\sigma^2$, mientras que $\mu$ suele ser desconocido.

### 2. El estadístico de prueba y su distribución
El fundamento analítico para este intervalo es el estadístico definido por la relación entre la varianza muestral y la poblacional:
$$\frac{(n-1)S^2}{\sigma^2}$$
Según el teorema fundamental presentado en las fuentes, esta variable aleatoria sigue una **distribución ji cuadrada ($\chi^2$)** con $\nu = n-1$ grados de libertad. A diferencia de la distribución normal o la $t$ de Student, la curva ji cuadrada no es simétrica y solo asume valores positivos, lo que requiere una notación específica para sus valores críticos.

### 3. Selección de valores críticos
Para un nivel de confianza deseado de $100(1-\alpha)\%$, se deben identificar dos puntos en el eje horizontal de la distribución ji cuadrada que capturen el área central:
*   **$\chi^2_{\alpha/2, n-1}$:** El valor crítico que deja un área de $\alpha/2$ en la cola superior (derecha).
*   **$\chi^2_{1-\alpha/2, n-1}$:** El valor crítico que deja un área de $1-\alpha/2$ a su derecha, lo que equivale a dejar $\alpha/2$ en la cola inferior (izquierda).

### 4. Derivación analítica del intervalo
El desarrollo matemático comienza con la siguiente proposición de probabilidad:
$$P\left( \chi^2_{1-\alpha/2, n-1} \le \frac{(n-1)S^2}{\sigma^2} \le \chi^2_{\alpha/2, n-1} \right) = 1 - \alpha$$.

Para aislar el parámetro $\sigma^2$ en el centro de la desigualdad, se realizan las siguientes manipulaciones algebraicas:
1.  Se toma el recíproco de cada término, lo que invierte el sentido de las desigualdades.
2.  Se multiplica toda la expresión por la cantidad constante $(n-1)S^2$.
3.  Al reordenar los términos para mantener el formato de límite inferior a la izquierda y límite superior a la derecha, se obtiene la expresión final.

Para resolver tu tarea basándonos en el material de estudio de Devore (Capítulos 7 y 9), a continuación se presenta la justificación analítica y conceptual de la distribución de la varianza muestral.

### 1. El Teorema Fundamental
Según el texto de Devore, si se extrae una muestra aleatoria $X_1, \dots, X_n$ de una población que sigue una **distribución normal** con parámetros $\mu$ y $\sigma^2$, se establece el siguiente teorema fundamental para la inferencia sobre la variabilidad:

La variable aleatoria definida por la relación entre la varianza muestral ($S^2$) y la poblacional ($\sigma^2$):
$$\frac{(n-1)S^2}{\sigma^2} = \frac{\sum_{i=1}^{n}(X_i - \bar{X})^2}{\sigma^2}$$
tiene una distribución de probabilidad **ji cuadrada ($\chi^2$)**.

### 2. Justificación Analítica de la Expresión
La igualdad que planteas en tu consulta se deriva directamente de la definición de la varianza muestral ($S^2$):
*   **Definición de $S^2$:** Se define como $S^2 = \frac{1}{n-1} \sum_{i=1}^{n}(X_i - \bar{X})^2$.
*   **Aislamiento de la sumatoria:** Al multiplicar ambos lados por $(n-1)$, obtenemos $(n-1)S^2 = \sum_{i=1}^{n}(X_i - \bar{X})^2$.
*   **Estandarización:** Al dividir toda la expresión por la varianza poblacional $\sigma^2$, resulta la variable que estudiamos: $\frac{(n-1)S^2}{\sigma^2}$.

### 3. ¿Con cuántos grados de libertad?
La distribución ji cuadrada está regida por un único parámetro denominado **grados de libertad**, denotado frecuentemente con la letra griega $\nu$ (nu).

Para este estadístico en particular, el número de grados de libertad es **$n - 1$**.

#### Justificación de los grados de libertad ($n-1$)
El texto explica que el uso de $n-1$ en lugar de $n$ no es arbitrario:
*   **Restricción matemática:** Aunque el cálculo de $S^2$ se basa en $n$ desviaciones individuales ($X_i - \bar{X}$), estas cantidades están sujetas a la restricción de que su suma debe ser obligatoriamente cero ($\sum (X_i - \bar{X}) = 0$).
*   **Valores libres:** Debido a esta restricción, si conocemos los valores de $n-1$ desviaciones, el valor de la última desviación queda determinado automáticamente para que la suma sea cero. Por lo tanto, solo existen $n-1$ desviaciones "libremente determinadas" en las que se basa la estimación de la varianza.
*   **Insesgamiento:** El uso del divisor $n-1$ permite que $S^2$ sea un **estimador insesgado** de $\sigma^2$. Si se utilizara $n$ en el denominador, el estimador tendería a subestimar la varianza poblacional real.

### 4. Importancia en el Capítulo 7
Esta relación es la "piedra angular" para construir **intervalos de confianza para la varianza ($\sigma^2$)**. Al saber que el estadístico sigue una distribución ji cuadrada con $n-1$ grados de libertad, podemos utilizar los valores críticos de las tablas ($\chi^2_{\alpha/2, n-1}$ y $\chi^2_{1-\alpha/2, n-1}$) para acotar el valor de $\sigma^2$ con un nivel de confianza determinado.

Para calcular un límite de confianza superior (LCS) para la desviación estándar poblacional en una población normal, utilizaremos la distribución **ji cuadrada ($\chi^2$)** con $n-1$ grados de libertad.

*   Tamaño de la muestra ($n$): 15.
*   Grados de libertad ($\nu$): $n - 1 = 14$.
*   Suma de los datos ($\sum x_i$): $379.7$.
*   Suma de los cuadrados ($\sum x_i^2$): $9647.37$.

**Paso 2: Cálculo de la varianza muestral ($s^2$)**
Utilizando la fórmula de cómputo para el numerador $(n-1)s^2$:
$$(n-1)s^2 = \sum x_i^2 - \frac{(\sum x_i)^2}{n}$$
$$(n-1)s^2 = 9647.37 - \frac{(379.7)^2}{15} = 9647.37 - 9611.4727 = \mathbf{35.8973}$$
Entonces, la varianza muestral es $s^2 = 35.8973 / 14 \approx 2.5641$ y la desviación estándar muestral es $s \approx 1.601$.

**Paso 3: Selección del valor crítico de $\chi^2$**
Para un límite de confianza **unilateral superior** del 95% ($1-\alpha = 0.95$), necesitamos el valor crítico que deje el 95% del área a su derecha (o el 5% en la cola inferior).
*   Buscamos en la **Tabla A.7** el valor de $\chi^2_{1-\alpha, \nu}$, que es $\chi^2_{0.95, 14}$.
*   En la fila 14 y columna 0.95, el valor es **6.571**.

La expresión analítica para el LCS de la varianza $\sigma^2$ es:

Sustituyendo los valores:
$$\sigma^2 \le \frac{35.8973}{6.571} \approx 5.463$$

$$\sigma \le \sqrt{5.463} \approx \mathbf{2.337}$$

>
> **c. Sean $x_{(2)}$ la segunda más pequeña de las $x_i$ y $x_{(n-1)}$ la segunda más grande de las $x_i$. ¿Cuál es el coeficiente de confianza del intervalo $(x_{(2)}, x_{(n-1)})$ para $\tilde{\mu}$?**

### Solución Paso a Paso

#### Parte a: Demostración Analítica
Para demostrar la probabilidad de que el intervalo formado por los valores extremos capture a la mediana, utilizaremos el enfoque del complemento sugerido en el texto.

1.  **Definición del complemento:** El evento de que la mediana **no** esté en el intervalo ocurre si todos los valores de la muestra son menores que la mediana o si todos son mayores que ella.
    *   Evento $A$: $\max(X_i) < \tilde{\mu}$ (Todos los valores son menores que la mediana).
    *   Evento $B$: $\min(X_i) > \tilde{\mu}$ (Todos los valores son mayores que la mediana).
2.  **Cálculo de probabilidades individuales:** Como las observaciones son independientes y la probabilidad de estar por encima o por debajo de la mediana es $0.5$:
    *   $P(A) = P(X_1 < \tilde{\mu} \cap X_2 < \tilde{\mu} \cap \dots \cap X_n < \tilde{\mu}) = (0.5)^n$.
    *   $P(B) = P(X_1 > \tilde{\mu} \cap X_2 > \tilde{\mu} \cap \dots \cap X_n > \tilde{\mu}) = (0.5)^n$.
3.  **Probabilidad del complemento:** Dado que $A$ y $B$ son mutuamente excluyentes (no pueden ser todos menores y todos mayores a la vez):
    *   $P(\text{Complemento}) = (0.5)^n + (0.5)^n = 2 \cdot (1/2)^n = (1/2)^{n-1}$.
4.  **Conclusión:** Por lo tanto, la probabilidad del evento original es:
    *   $P(\min(X_i) \le \tilde{\mu} \le \max(X_i)) = 1 - (1/2)^{n-1}$.

#### Parte b: Aplicación Práctica (Alanina en infantes)
1.  **Datos ordenados:** Para identificar los valores extremos, ordenamos la muestra de $n=6$ infantes:
    *   $1.44, 2.70, 2.80, 2.84, 2.94, 3.54$.
2.  **Verificación del nivel de confianza:** Con $n=6$, el nivel de confianza es:
    *   $1 - (1/2)^{6-1} = 1 - (1/2)^5 = 1 - 1/32 = 1 - 0.03125 = 0.96875 \approx 97\%$.
3.  **Resultado:** El intervalo de confianza de aproximadamente 97% para la mediana de alanina es **(1.44, 3.54)** mg/100 ml.

#### Parte c: Coeficiente de Confianza para el Segundo Rango
Para el intervalo $(x_{(2)}, x_{(n-1)})$, la mediana queda fuera si la muestra tiene $0$ o $1$ observación menor que $\tilde{\mu}$, o bien $0$ o $1$ observación mayor que $\tilde{\mu}$.

1.  **Modelo Binomial:** Sea $Y$ el número de observaciones menores que la mediana. $Y$ sigue una distribución binomial con $p=0.5$ y $n$ ensayos.
2.  **Probabilidad de falla:** La mediana no estará en el intervalo si $Y \in \{0, 1, n-1, n\}$.
    *   $P(Y=0) = \binom{n}{0}(0.5)^n = (0.5)^n$
    *   $P(Y=1) = \binom{n}{1}(0.5)^n = n(0.5)^n$
    *   $P(Y=n-1) = n(0.5)^n$
    *   $P(Y=n) = (0.5)^n$
3.  **Suma de probabilidades de falla:** $P(\text{falla}) = (2 + 2n)(0.5)^n = 2(n+1)(1/2)^n = (n+1)(1/2)^{n-1}$.
4.  **Coeficiente de confianza final:**
    *   $1 - (n+1)(1/2)^{n-1}$.

Este método es un procedimiento "libre de distribución" porque su validez no depende de que la población sea normal, solo de que sea continua.

Para demostrar que las áreas de las colas en una distribución ji cuadrada ($\chi^2$) se dividen equitativamente como $A_1 = A_2 = \alpha/2$ al construir un intervalo de confianza, debemos remitirnos a la definición de los valores críticos y a la geometría de la densidad de probabilidad presentada en el texto de Devore.

### 1. El principio del área total
De acuerdo con las propiedades fundamentales de cualquier función de densidad de probabilidad continua, el área total bajo la curva debe ser igual a 1. En su planteamiento, usted define el área total como:
$$A = A_1 + (1 - \alpha) + A_2 = 1$$
Donde:
*   $(1 - \alpha)$ es el **nivel de confianza** (el área central que deseamos capturar).
*   $\alpha$ es el **nivel de significancia** o el riesgo total de error, que se distribuye en las colas.

### 2. Definición de la notación de valores críticos
El texto establece una convención específica para identificar puntos en el eje horizontal de la distribución ji cuadrada:
*   **$\chi^2_{\alpha, \nu}$:** Se define como el valor crítico ji cuadrada de modo que el área $\alpha$ bajo la curva con $\nu$ grados de libertad queda a la **derecha** de dicho valor.
*   Para capturar un área central de $1 - \alpha$, el procedimiento estándar requiere descartar una porción del área total en ambos extremos de la distribución asimétrica.

### 3. Demostración de la división equitativa ($A_1 = A_2 = \alpha/2$)
Para aislar el parámetro de interés (como la varianza $\sigma^2$) en un intervalo central, se seleccionan dos valores críticos que actúan como fronteras:

*   **Límite Superior del estadístico (Cola Derecha):** Se elige el valor $\chi^2_{\alpha/2, \nu}$. Por definición, este valor deja un área de exactamente $\alpha/2$ a su derecha. Por lo tanto, el área de la cola superior es **$A_2 = \alpha/2$**.
*   **Límite Inferior del estadístico (Cola Izquierda):** Se elige el valor $\chi^2_{1-\alpha/2, \nu}$. Por definición, este valor deja un área de $1 - \alpha/2$ a su derecha.
    *   Dado que el área total es 1, el área a la **izquierda** de este punto es: $1 - (1 - \alpha/2) = \alpha/2$.
    *   Por lo tanto, el área de la cola inferior es **$A_1 = \alpha/2$**.

### 4. Verificación de la suma
Si sumamos las áreas definidas analíticamente por los valores críticos:
1.  **Suma de las colas:** $A_1 + A_2 = \alpha/2 + \alpha/2 = \alpha$. Esto cumple con su condición inicial.
2.  **Área total:** $A_1 + (1-\alpha) + A_2 = \alpha/2 + 1 - \alpha + \alpha/2 = 1$. Esto cumple con la propiedad de las distribuciones de probabilidad.

### Conclusión técnica
Esta división equitativa es la que fundamenta la **Ecuación (7.17)** del libro, la cual expresa la probabilidad de que el estadístico ji cuadrada se encuentre entre ambos límites:
$$P\left( \chi^2_{1-\alpha/2, n-1} \le \frac{(n-1)S^2}{\sigma^2} \le \chi^2_{\alpha/2, n-1} \right) = 1 - \alpha$$.

Al utilizar $\alpha/2$ en cada extremo, el investigador garantiza que el intervalo de confianza sea "centrado" en términos de probabilidad, a pesar de que la curva ji cuadrada sea asimétrica y no esté centrada en el eje horizontal.

### 1. Definición de Consistencia
De acuerdo con el enunciado del problema y los conceptos de estimación puntual, se dice que un estimador $\hat{\theta}$ es **consistente** si, para cualquier número arbitrario $\epsilon > 0$, la probabilidad de que el estimador se aleje del valor verdadero del parámetro en una cantidad mayor o igual a $\epsilon$ tiende a cero a medida que el tamaño de la muestra ($n$) crece hacia el infinito,.

Analíticamente, $\hat{\theta}$ es consistente para $\theta$ si:
$$\lim_{n \to \infty} P(|\hat{\theta} - \theta| \geq \epsilon) = 0$$.

### 2. Fundamento Matemático: Desigualdad de Chebyshev
Para realizar esta demostración, el problema instruye utilizar la desigualdad de Chebyshev. Según el texto de Devore, para cualquier variable aleatoria $Y$ con media $\mu_Y$ y varianza $\sigma_Y^2$, y para cualquier número $\epsilon > 0$, la desigualdad establece un límite superior para la probabilidad de desviación:
$$P(|Y - \mu_Y| \geq \epsilon) \leq \frac{\sigma_Y^2}{\epsilon^2}$$,,.

Esta herramienta permite acotar qué tan probable es que una variable aleatoria se encuentre lejos de su valor esperado en términos de su varianza.

### 3. Aplicación a la Media Muestral ($\bar{X}$)
Para demostrar que $\bar{X}$ es consistente para $\mu$, identificamos a $Y$ con el estadístico $\bar{X}$. Primero, debemos determinar las propiedades de la distribución muestral de $\bar{X}$:
*   **Valor esperado:** El estimador $\bar{X}$ es un estimador insesgado de la media poblacional, por lo que $E(\bar{X}) = \mu$,.
*   **Varianza:** La varianza de la media muestral disminuye conforme aumenta el tamaño de la muestra y se define como $V(\bar{X}) = \sigma_{\bar{X}}^2 = \frac{\sigma^2}{n}$,.

Sustituyendo estos valores en la expresión de Chebyshev:
$$P(|\bar{X} - \mu| \geq \epsilon) \leq \frac{V(\bar{X})}{\epsilon^2}$$
$$P(|\bar{X} - \mu| \geq \epsilon) \leq \frac{\sigma^2/n}{\epsilon^2}$$
$$P(|\bar{X} - \mu| \geq \epsilon) \leq \frac{\sigma^2}{n \cdot \epsilon^2}$$,.

### 4. Demostración mediante el Límite
El último paso consiste en evaluar el comportamiento de esta probabilidad cuando el tamaño de la muestra tiende al infinito ($n \to \infty$):
$$\lim_{n \to \infty} P(|\bar{X} - \mu| \geq \epsilon) \leq \lim_{n \to \infty} \frac{\sigma^2}{n \cdot \epsilon^2}$$

Dado que $\sigma^2$ es una constante finita ($\sigma^2 < \infty$) y $\epsilon^2$ es una constante positiva fija, el denominador $n \cdot \epsilon^2$ crece sin límite mientras el numerador permanece constante. Por lo tanto:
$$\lim_{n \to \infty} \frac{\sigma^2}{n \cdot \epsilon^2} = 0$$

### 5. Conclusión
Como se ha demostrado que la probabilidad de que la media muestral $\bar{X}$ difiera de la media verdadera $\mu$ en más de $\epsilon$ tiende a cero cuando $n$ aumenta, se cumple la definición formal de consistencia,. 

**Resultado final:** La media muestral $\bar{X}$ es un estimador consistente de la media poblacional $\mu$ siempre que la varianza poblacional sea finita.

# Ejemplo

$X:$ Cantidad de Hidrógeno neutro en galaxias espirales
n=10 galaxias encontraron que $\bar{x}=21.9$

Calcule el intervalo de confianza para $\mu$

Para desarrollar este ejemplo basado en los principios de inferencia estadística del Capítulo 7 de Devore, supongamos que un equipo de astrónomos estudia la masa de hidrógeno neutro ($X$) en una muestra de galaxias espirales para estimar el promedio poblacional ($\mu$).

A continuación, se desarrolla el ejemplo paso a paso utilizando la metodología de **Intervalos de Confianza para muestras pequeñas** (Caso III), ya que el tamaño de la muestra es $n < 40$ y no se conoce la desviación estándar poblacional ($\sigma$).

### 1. Definición del Problema y Datos
Se desea estimar, con un nivel de confianza del 95%, la cantidad promedio verdadera de hidrógeno neutro en un sector del universo.

*   **Variable aleatoria ($X$):** Masa de hidrógeno neutro (en unidades de $10^9$ masas solares).
*   **Tamaño de la muestra ($n$):** 10 galaxias.
*   **Media muestral ($\bar{x}$):** $21.9$.
*   **Desviación estándar muestral ($s$):** Supongamos, para completar el desarrollo ("etc..."), que los datos arrojaron una variabilidad de $s = 3.5$.
*   **Suposición indispensable:** Se debe suponer que la cantidad de hidrógeno en estas galaxias sigue una **distribución normal**.

### 2. Identificación del Método (Caso III)
Dado que $n=10$ (muestra pequeña) y $\sigma$ es desconocida, las fuentes establecen que el estadístico estandarizado:
$$T = \frac{\bar{X} - \mu}{S/\sqrt{n}}$$
sigue una **distribución $t$ de Student** con $\nu = n - 1$ grados de libertad.

### 3. Determinación de Parámetros Críticos
1.  **Grados de libertad ($\nu$):**
    $$\nu = 10 - 1 = 9 \text{}$$
2.  **Nivel de confianza ($1-\alpha$):** 95%, por lo tanto $\alpha = 0.05$ y el área en cada cola es $\alpha/2 = 0.025$.
3.  **Valor crítico ($t_{\alpha/2, \nu}$):** Consultando la **Tabla A.5** (Valores críticos para distribuciones $t$), en la intersección de la fila $\nu = 9$ y la columna $0.025$, obtenemos:
    $$t_{0.025, 9} = \mathbf{2.262} \text{}$$

### 4. Cálculo del Intervalo de Confianza (IC)
La fórmula general para el intervalo es:
$$IC = \bar{x} \pm t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}} \text{}$$

Sustituyendo los valores:
1.  **Error estándar estimado:**
    $$\frac{s}{\sqrt{n}} = \frac{3.5}{\sqrt{10}} \approx \frac{3.5}{3.162} \approx 1.1068$$
2.  **Margen de error ($E$):**
    $$E = 2.262 \cdot 1.1068 \approx 2.5036$$
3.  **Límites del intervalo:**
    *   Límite Inferior: $21.9 - 2.50 = 19.4$
    *   Límite Superior: $21.9 + 2.50 = 24.4$

### 5. Resultado e Interpretación
El intervalo de confianza del 95% para la masa promedio verdadera de hidrógeno neutro es **(19.4, 24.4)**.

**Interpretación técnica:** Tenemos un 95% de confianza de que la masa promedio real de hidrógeno neutro en todas las galaxias espirales de este tipo se encuentra entre 19.4 y 24.4 unidades. Si este experimento de muestreo de 10 galaxias se repitiera muchas veces, el 95% de los intervalos resultantes capturarían el valor real de $\mu$.

*(Nota: Este ejemplo ha sido desarrollado siguiendo la estructura técnica y los valores críticos de las tablas de las fuentes consultadas, como la Tabla A.5 y la metodología del Caso III de la página 292 del texto).*

Ejemplo 7.12 del libro

Para aplicar los conceptos de inferencia a la **predicción**, debemos pasar de estimar un parámetro poblacional (como la media $\mu$) a estimar el valor de una **sola observación futura** ($X_{n+1}$). En el contexto del ejemplo de las galaxias espirales, esto significa que ya no buscamos el rango donde se encuentra el promedio de todas las galaxias, sino el rango donde probablemente caerá la masa de hidrógeno de **una galaxia específica** que observemos en el futuro.

A continuación se detalla cómo se aplica este método, denominado **Intervalo de Predicción (IP)**, utilizando los datos del mensaje anterior:

### 1. Diferencia Conceptual
Mientras que el intervalo de confianza (IC) se contrae hacia el valor de la media a medida que aumenta el tamaño de la muestra, el intervalo de predicción (IP) siempre será más ancho. Esto se debe a que el IP debe considerar dos fuentes de incertidumbre:
1.  La incertidumbre sobre la ubicación de la media poblacional $\mu$.
2.  La variabilidad inherente de la propia observación individual alrededor de esa media.

### 2. Fórmula del Intervalo de Predicción
Para una población normal con $\sigma$ desconocida y muestra pequeña, la fórmula para un nivel de confianza de $100(1-\alpha)\%$ es:
$$IP = \bar{x} \pm t_{\alpha/2, n-1} \cdot s\sqrt{1 + \frac{1}{n}}$$

### 3. Aplicación al Ejemplo de Hidrógeno en Galaxias
Retomando los datos previos:
*   **Muestra ($n$):** 10 galaxias.
*   **Media muestral ($\bar{x}$):** 21.9.
*   **Desviación estándar muestral ($s$):** 3.5.
*   **Grados de libertad ($\nu$):** $n - 1 = 9$.
*   **Valor crítico ($t_{0.025, 9}$):** 2.262 (para un 95% de confianza).

#### Paso a paso del cálculo:
1.  **Cálculo del error estándar de la predicción:**
    A diferencia del IC, aquí el error se multiplica por un factor que incluye el "$1$" extra bajo la raíz:
    $$s\sqrt{1 + 1/10} = 3.5\sqrt{1.1} \approx 3.5 \cdot 1.0488 \approx 3.6708$$
2.  **Margen de error ($E_{pred}$):**
    $$E_{pred} = 2.262 \cdot 3.6708 \approx 8.303$$
3.  **Límites del intervalo:**
    *   Límite Inferior: $21.9 - 8.30 = 13.60$
    *   Límite Superior: $21.9 + 8.30 = 30.20$

### 4. Resultado e Interpretación
El **Intervalo de Predicción del 95%** para la masa de hidrógeno de una nueva galaxia seleccionada al azar es **(13.60, 30.20)**.

**Interpretación Técnica:** Tenemos un 95% de confianza en que el valor de la masa de hidrógeno para **la próxima galaxia** que se observe se encontrará entre 13.60 y 30.20 unidades. Si repitiéramos este proceso de muestreo y predicción muchas veces, a la larga el 95% de los intervalos calculados contendrían el valor real observado de la nueva galaxia. 

Note que este intervalo es mucho más amplio que el de la media (que era 19.4 a 24.4), reflejando la mayor dificultad de predecir un solo evento en comparación con estimar un promedio.
