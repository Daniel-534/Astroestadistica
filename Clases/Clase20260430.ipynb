El **poder de la prueba** (o potencia) se define analíticamente como la probabilidad de rechazar la hipótesis nula ($H_0$) cuando esta es falsa. Es una medida de la capacidad de un procedimiento estadístico para detectar correctamente un alejamiento del valor nulo.

Los aspectos fundamentales del poder de la prueba según las fuentes son:

### 1. Relación con el Error de Tipo II ($\beta$)
El poder es el complemento de la probabilidad de cometer un error de tipo II:
$$\text{Potencia} = 1 - \beta$$
Mientras que $\beta$ representa la probabilidad de no rechazar una $H_0$ falsa, la potencia es la probabilidad de tomar la decisión correcta y rechazarla.

### 2. Factores que aumentan el poder
El poder de una prueba no es un valor único, sino que depende de varias condiciones del experimento:
*   **Magnitud de la diferencia:** La potencia aumenta a medida que el verdadero valor del parámetro se aleja del valor nulo ($\mu_0$); es más fácil detectar desviaciones grandes que pequeñas.
*   **Tamaño de la muestra ($n$):** Incrementar el tamaño de la muestra reduce el error estándar, lo que disminuye $\beta$ y, por consecuencia, aumenta la potencia.
*   **Variabilidad ($\sigma$):** Una desviación estándar poblacional menor resulta en una potencia más alta, ya que los datos son menos ruidosos y las diferencias son más fáciles de identificar.
*   **Nivel de significancia ($\alpha$):** Existe un compromiso entre los errores; si se reduce $\alpha$ (haciendo la prueba más conservadora contra errores de tipo I), la región de rechazo se hace más pequeña, lo que aumenta $\beta$ y reduce la potencia.

### 3. Significación estadística vs. práctica
Es crítico notar que con tamaños de muestra muy grandes, el poder de la prueba puede ser tan alto que detecte desviaciones minúsculas de la hipótesis nula que no tienen ninguna importancia o **significación práctica** en el mundo real.

### 4. Herramientas de cálculo
Para evaluar el poder, se utilizan comúnmente **curvas de potencia** (gráficas que muestran la potencia en función de la diferencia a detectar) o software estadístico especializado. En pruebas complejas como ANOVA, el poder depende de un "parámetro de no centralidad" que relaciona la variabilidad de los efectos con la varianza del error.

Las pruebas de hipótesis con **dos muestras** se utilizan para comparar parámetros (como medias, proporciones o varianzas) de dos poblaciones distintas basándose en datos muestrales independientes o apareados.

A continuación se resumen los casos principales detallados en la literatura técnica:

### 1. Comparación de dos medias poblacionales ($\mu_1 - \mu_2$)
Este procedimiento evalúa si existe una diferencia significativa entre los promedios de dos grupos.

*   **Caso I: Varianzas conocidas y poblaciones normales:** Se utiliza el estadístico **$Z$**. La hipótesis nula suele ser $H_0: \mu_1 - \mu_2 = \Delta_0$ (donde $\Delta_0$ es a menudo 0).
*   **Caso II: Muestras grandes ($m, n > 40$):** Por el Teorema del Límite Central, se utiliza el estadístico **$Z$** incluso si las poblaciones no son normales, sustituyendo las varianzas poblacionales por las muestrales ($s_1^2, s_2^2$).
*   **Caso III: Muestras pequeñas y varianzas desconocidas:** Se asume normalidad y se utiliza la **prueba $t$ de dos muestras**. El número de grados de libertad se estima mediante la aproximación de Welch-Satterthwaite si las varianzas se consideran desiguales.

### 2. Análisis de datos apareados
Se aplica cuando las observaciones de ambas muestras están relacionadas entre sí (por ejemplo, mediciones antes y después en el mismo sujeto).
*   Se calculan las diferencias individuales ($d_i = x_i - y_i$).
*   Se realiza una **prueba $t$ para una sola muestra** sobre estas diferencias para probar $H_0: \mu_D = \Delta_0$.

### 3. Comparación de dos proporciones ($\hat{p}_1 - \hat{p}_2$)
Se utiliza para determinar si la proporción de "éxitos" es la misma en dos poblaciones independientes.
*   **Muestras grandes:** Se utiliza un estadístico **$Z$** basado en una proporción combinada ($\hat{p}$) de ambas muestras bajo la suposición de que $H_0: p_1 = p_2$ es verdadera.

### 4. Comparación de dos varianzas ($\sigma_1^2 / \sigma_2^2$)
Este test es útil para verificar si dos poblaciones tienen la misma dispersión.
*   **Estadístico $F$:** Se calcula como la razón de las varianzas muestrales ($f = s_1^2 / s_2^2$).
*   La distribución nula sigue una **distribución $F$** con $m-1$ y $n-1$ grados de libertad.

### Resumen de Estadísticos de Prueba
| Parámetro | Caso | Estadístico |
| :--- | :--- | :--- |
| Medias ($\mu_1, \mu_2$) | Varianzas conocidas | $Z = \frac{(\bar{x}-\bar{y}) - \Delta_0}{\sqrt{\sigma_1^2/m + \sigma_2^2/n}}$ |
| Medias ($\mu_1, \mu_2$) | Muestras pequeñas / $s$ desconocida | $t = \frac{(\bar{x}-\bar{y}) - \Delta_0}{\sqrt{s_1^2/m + s_2^2/n}}$ |
| Proporciones ($p_1, p_2$) | Muestras grandes | $Z = \frac{\hat{p}_1 - \hat{p}_2}{\sqrt{\hat{p}\hat{q}(1/m + 1/n)}}$ |
| Varianzas ($\sigma_1^2, \sigma_2^2$) | Poblaciones normales | $F = s_1^2 / s_2^2$ |

### Ejemplo 9.2: Longevidad y Afiliación Profesional

**Contexto y Datos:**
Se analizó una muestra de **215 médicos** graduados de Harvard que murieron entre 1974 y 1977. El objetivo es determinar si los médicos en práctica privada (equivalente al grupo en la "industria") viven, en promedio, más tiempo después de graduarse que aquellos con carreras académicas.

*   **Grupo 1 (Práctica de tiempo completo):** $m = 125$ médicos. Vivieron un promedio de $\bar{x} = 48.9$ años tras su graduación.
*   **Grupo 2 (Afiliaciones académicas):** $n = 90$ médicos. Vivieron un promedio de $\bar{y} = 43.2$ años tras su graduación.
*   **Suposiciones de varianza:** El ejemplo utiliza valores ilustrativos de $\sigma_1 = 14.6$ y $\sigma_2 = 14.4$.

**Desarrollo Analítico:**

1.  **Parámetros:** $\mu_1$ es el promedio de años vividos tras la graduación para práctica privada y $\mu_2$ para práctica académica.
2.  **Hipótesis:** 
    *   $H_0: \mu_1 - \mu_2 = 0$ (No hay diferencia).
    *   $H_a: \mu_1 - \mu_2 > 0$ (La vida media en práctica privada es mayor).
3.  **Estadístico de prueba ($z$):** Se calcula midiendo la diferencia de medias observada contra el error estándar:
    $$z = \frac{48.9 - 43.2}{\sqrt{\frac{14.6^2}{125} + \frac{14.4^2}{90}}} \approx 2.85$$.
4.  **Valor P:** Para una prueba de cola superior, el valor P es $1 - \Phi(2.85) = 0.0022$.

**Conclusión:**
Como el valor P ($0.0022$) es menor que el nivel de significancia estándar ($\alpha = 0.01$), **se rechaza la hipótesis nula**. Los datos sugieren fuertemente que, en esta población, los médicos dedicados a la práctica de tiempo completo vivieron significativamente más años después de su graduación que sus colegas en la academia.
