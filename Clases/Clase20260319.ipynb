### 1. Proposición de Valor Esperado para una Función $h(X, Y)$

La proposición que aparece en la **página 197** (dentro de la sección 5.2) establece las reglas para calcular el valor esperado de una función que depende de dos variables aleatorias conjuntamente distribuidas.

**Enunciado de la Proposición:**
Sean $X$ y $Y$ variables aleatorias conjuntamente distribuidas con función masa de probabilidad $p(x, y)$ (si son discretas) o función de densidad de probabilidad $f(x, y)$ (si son continuas). Entonces, el **valor esperado de una función $h(X, Y)$**, denotado por $E[h(X, Y)]$ o $\mu_{h(X, Y)}$, se calcula de la siguiente manera:

*   **Si $X$ y $Y$ son discretas:**
    $$E[h(X, Y)] = \sum_x \sum_y h(x, y) \cdot p(x, y)$$
*   **Si $X$ y $Y$ son continuas:**
    $$E[h(X, Y)] = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} h(x, y) \cdot f(x, y) \, dx \, dy$$

Esta proposición permite hallar el promedio a largo plazo de una función de dos variables sin necesidad de obtener primero la distribución de probabilidad de la propia función.

### 2. Ejercicio de Astronomía (Adaptación del Ejercicio 26, Sección 5.2)

Para ilustrar este concepto, tomaremos el **ejercicio 26** de la página 202 (que originalmente trata sobre un transbordador que transporta carros y autobuses) y lo modificaremos para un contexto astronómico.

**Enunciado Adaptado:**
Un telescopio espacial realiza una serie de exposiciones automáticas para detectar objetos celestes. En cada ciclo de observación, el telescopio puede identificar dos tipos de objetos: **Cúmulos Estelares ($X$)** y **Agujeros Negros de masa estelar ($Y$)**. 

A cada hallazgo se le asigna un "Puntaje de Relevancia Científica": cada cúmulo estelar aporta **3 puntos** y cada agujero negro aporta **10 puntos**. El número de objetos detectados por ciclo sigue la distribución de probabilidad conjunta presentada en la tabla del ejercicio 7:

| **0 (Cúmulos)** | 0.025 | 0.015 | 0.010 |
| **1 (Cúmulo)** | 0.050 | 0.030 | 0.020 |

**Tarea:** Calcule el **puntaje de relevancia científica esperado** para un ciclo de observación individual.

### 3. Resolución y Aplicación de la Proposición

Para resolver el problema, definimos la función de puntaje total como **$h(X, Y) = 3X + 10Y$**. Aplicaremos la proposición de valor esperado para variables discretas.

#### Paso 1: Obtención de las Probabilidades Marginales
Para facilitar el cálculo, primero sumamos las probabilidades de las filas para $X$ (Cúmulos) y las columnas para $Y$ (Agujeros Negros):

*   **Para $X$ (Cúmulos):**
    *   $p_X(0) = 0.025 + 0.015 + 0.010 = 0.050$
    *   $p_X(1) = 0.050 + 0.030 + 0.020 = 0.100$
    *   $p_X(2) = 0.125 + 0.075 + 0.050 = 0.250$
    *   $p_X(3) = 0.150 + 0.090 + 0.060 = 0.300$
    *   $p_X(4) = 0.100 + 0.060 + 0.040 = 0.200$
    *(Nota: La suma total es 0.90 según los datos proporcionados en la fuente).*

*   **Para $Y$ (Agujeros Negros):**
    *   $p_Y(0) = 0.025 + 0.050 + 0.125 + 0.150 + 0.100 = 0.450$
    *   $p_Y(1) = 0.015 + 0.030 + 0.075 + 0.090 + 0.060 = 0.270$
    *   $p_Y(2) = 0.010 + 0.020 + 0.050 + 0.060 + 0.040 = 0.180$

#### Paso 2: Cálculo de los Valores Esperados Individuales
Calculamos la media de cada tipo de objeto detectado:
*   $E(X) = 0(0.05) + 1(0.10) + 2(0.25) + 3(0.30) + 4(0.20) = 0 + 0.10 + 0.50 + 0.90 + 0.80 = \mathbf{2.3}$
*   $E(Y) = 0(0.45) + 1(0.27) + 2(0.18) = 0 + 0.27 + 0.36 = \mathbf{0.63}$

#### Paso 3: Aplicación de la Proposición (Propiedad de Linealidad)
Utilizando la proposición para funciones lineales (donde $E[aX + bY] = aE(X) + bE(Y)$), calculamos el puntaje esperado:
$$E[3X + 10Y] = 3 \cdot E(X) + 10 \cdot E(Y)$$
$$E = 3(2.3) + 10(0.63) = 6.9 + 6.3 = \mathbf{13.2}$$

**Interpretación en Astronomía:**
El resultado de **13.2 puntos** indica que, si el telescopio realiza un gran número de ciclos de observación bajo estas condiciones, el promedio de puntos de relevancia científica obtenidos por ciclo tenderá a estabilizarse en 13.2. Esto permite a los directores de la misión científica estimar el rendimiento promedio del equipo a largo plazo.

A continuación, presento las definiciones y proposiciones solicitadas del texto de Devore, seguidas de su aplicación al ejercicio de astronomía desarrollado anteriormente.

### 1. Definición de Covarianza (Página 198)
La **covarianza** entre dos variables aleatorias $X$ y $Y$ es:
$$Cov(X, Y) = E[(X - \mu_X)(Y - \mu_Y)]$$
Esta se calcula como:
*   **Caso discreto:** $\sum_x \sum_y (x - \mu_X)(y - \mu_Y)p(x, y)$
*   **Caso continuo:** $\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} (x - \mu_X)(y - \mu_Y)f(x, y) dx dy$.

### 2. Proposición: Fórmula abreviada de la Covarianza (Página 199)
La siguiente fórmula simplifica los cálculos al no requerir sustracciones intermedias:
$$Cov(X, Y) = E(XY) - \mu_X \cdot \mu_Y$$.

### 3. Definición de Correlación (Página 200)
El **coeficiente de correlación** de $X$ y $Y$, denotado por $Corr(X, Y)$, $\rho_{X,Y}$ o simplemente $\rho$, está definido por:
$$\rho_{X,Y} = \frac{Cov(X, Y)}{\sigma_X \cdot \sigma_Y}$$.

### 4. Proposiciones sobre la Correlación (Página 200)

**Proposición 1:**
1.  Si $a$ y $c$ son ambas positivas o ambas negativas, $Corr(aX + b, cY + d) = Corr(X, Y)$.
2.  Para dos variables aleatorias cualesquiera $X$ y $Y$, $-1 \le Corr(X, Y) \le 1$.

**Proposición 2:**
1.  Si $X$ y $Y$ son independientes, entonces $\rho = 0$, pero $\rho = 0$ no implica independencia.
2.  $\rho = 1$ o $-1$ si y sólo si $Y = aX + b$ para algunos números $a$ y $b$ con $a \neq 0$.

### 5. Aplicación al ejercicio de Astronomía
Retomamos el ejemplo del telescopio que detecta **Cúmulos Estelares ($X$)** y **Agujeros Negros ($Y$)**. 

**Datos previos:**
*   $\mu_X = 2.3$
*   $\mu_Y = 0.63$
*   $\sigma_X = \sqrt{1.71} \approx 1.3077$
*   Para $\sigma_Y$, calculamos $E(Y^2) = 0^2(0.45) + 1^2(0.27) + 2^2(0.18) = 0.99$.
*   $V(Y) = 0.99 - (0.63)^2 = 0.5931 \Rightarrow \sigma_Y = \sqrt{0.5931} \approx 0.7701$.

**Paso 1: Cálculo de $E(XY)$**
Utilizando la tabla de probabilidad conjunta (omitiendo valores donde $x$ o $y$ son 0):
$E(XY) = (1)(1)(0.03) + (1)(2)(0.02) + (2)(1)(0.075) + (2)(2)(0.05) + (3)(1)(0.09) + (3)(2)(0.06) + (4)(1)(0.06) + (4)(2)(0.04)$
$E(XY) = 0.03 + 0.04 + 0.15 + 0.20 + 0.27 + 0.36 + 0.24 + 0.32 = \mathbf{1.61}$

**Paso 2: Aplicación de la Proposición de Covarianza (p. 199)**
$$Cov(X, Y) = 1.61 - (2.3)(0.63) = 1.61 - 1.449 = \mathbf{0.161}$$
Como el valor es positivo, existe una relación directa: en los ciclos donde se detectan más cúmulos, también hay una ligera tendencia a detectar más agujeros negros.

**Paso 3: Cálculo de la Correlación ($\rho$)**
$$\rho = \frac{0.161}{(1.3077)(0.7701)} = \frac{0.161}{1.007} \approx \mathbf{0.160}$$

**Interpretación final:**
De acuerdo con las reglas de la página 200, dado que $|\rho| \le 0.5$, la relación lineal entre la detección de cúmulos y agujeros negros se clasifica como **débil**. Aunque hay una dependencia (no son independientes porque $Cov \neq 0$), la presencia de un objeto no permite predecir con fuerza la presencia del otro de forma lineal.

## Caso continuo

Para ilustrar el concepto de **valor esperado para variables continuas conjuntamente distribuidas**, adaptaremos el **Ejercicio 27 de la Sección 5.2** (página 202 del libro de Devore) a un contexto astronómico.

### 1. Proposición de Valor Esperado (Página 197)

La proposición fundamental para este cálculo establece que si $X$ y $Y$ son variables aleatorias continuas con una función de densidad de probabilidad (fdp) conjunta $f(x, y)$, el **valor esperado de cualquier función $h(X, Y)$** se calcula mediante la integral doble sobre todo el plano $xy$:

$$E[h(X, Y)] = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} h(x, y) \cdot f(x, y) \, dx \, dy$$

---

### 2. Adaptación al campo de la Astronomía

Dos telescopios robóticos, el **Telescopio Alfa** y el **Telescopio Beta**, están programados para iniciar una observación de una kilonova en el momento en que reciban una alerta de ondas gravitacionales. El tiempo de inicio de cada telescopio (medido como una fracción de la primera hora tras la alerta, $0 \le x, y \le 1$) es incierto y sigue distribuciones marginales distintas debido a sus diferentes sistemas de apuntado.

$X=$ Tiempo de inicio de la observación del telescopio alfa

$Y=$ Tiempo de inicio de la observación del telescopio beta

Supongamos que los tiempos de inicio son **independientes** y tienen las siguientes funciones de densidad:
*   $f_X(x) = 3x^2$ para $0 \le x \le 1$
*   $f_Y(y) = 2y$ para $0 \le y \le 1$

**Tarea:** Calcule el **tiempo esperado de desfase** entre el inicio de ambos telescopios. **Hint:** $h(X, Y) = |X - Y|$.

---

### 3. Resolución y Aplicación Matemática

#### Paso 1: Obtener la fdp conjunta
Como los tiempos de inicio son independientes, la función de densidad conjunta es el producto de las marginales:
$$f(x, y) = f_X(x) \cdot f_Y(y) = (3x^2)(2y) = 6x^2y$$
para $0 \le x \le 1$ y $0 \le y \le 1$.

#### Paso 2: Plantear la integral del valor esperado
Aplicando la proposición de la página 197:
$$E[|X - Y|] = \int_{0}^{1} \int_{0}^{1} |x - y| \cdot (6x^2y) \, dy \, dx$$

#### Paso 3: Resolver la integral dividiendo la región
Debido al valor absoluto $|x - y|$, debemos dividir la región de integración en dos partes: donde $x > y$ (desfase $x - y$) y donde $y > x$ (desfase $y - x$).

1.  **Región 1 ($y \le x$):**
    $$\int_{0}^{x} (x - y) 6x^2y \, dy = \int_{0}^{x} (6x^3y - 6x^2y^2) \, dy$$
    $$= [3x^3y^2 - 2x^2y^3]_0^x = 3x^5 - 2x^5 = \mathbf{x^5}$$

2.  **Región 2 ($y > x$):**
    $$\int_{x}^{1} (y - x) 6x^2y \, dy = \int_{x}^{1} (6x^2y^2 - 6x^3y) \, dy$$
    $$= [2x^2y^3 - 3x^3y^2]_x^1 = (2x^2 - 3x^3) - (2x^5 - 3x^5) = \mathbf{2x^2 - 3x^3 + x^5}$$

#### Paso 4: Integral externa respecto a $x$
Sumamos ambos resultados e integramos de 0 a 1:
$$E[|X - Y|] = \int_{0}^{1} (x^5 + 2x^2 - 3x^3 + x^5) \, dx = \int_{0}^{1} (2x^5 - 3x^3 + 2x^2) \, dx$$
$$E = \left[ \frac{2x^6}{6} - \frac{3x^4}{4} + \frac{2x^3}{3} \right]_0^1 = \frac{1}{3} - \frac{3}{4} + \frac{2}{3}$$
$$E = 1 - 0.75 = \mathbf{0.25}$$

### Interpretación Astronómica
El resultado de **0.25** indica que, bajo este modelo de probabilidad, el **desfase promedio esperado** entre el inicio de las observaciones de ambos telescopios es de **15 minutos** (0.25 de una hora). Esta información es crucial para la sincronización de datos multimensajero, permitiendo a los astrónomos prever cuánta información inicial del evento podría perderse o desfasarse entre los distintos instrumentos de la red global.

Para aplicar los conceptos de **covarianza** y **correlación** al ejercicio de los telescopios Alfa y Beta (basado en el modelo de variables continuas de nuestra conversación anterior), utilizaremos las definiciones y fórmulas fundamentales del libro de Devore.

### 1. Definiciones Teóricas (Fuente: Devore)

*   **Covarianza ($Cov(X, Y)$):** Es el valor esperado del producto de las desviaciones de las variables respecto a sus medias. Se utiliza la fórmula abreviada para simplificar los cálculos:
    $$Cov(X, Y) = E(XY) - \mu_X \cdot \mu_Y$$.
*   **Correlación ($\rho$):** Mide el grado de asociación lineal entre $X$ y $Y$. Se define como:
    $$\rho = \frac{Cov(X, Y)}{\sigma_X \cdot \sigma_Y}$$.
*   **Propiedad de Independencia:** Según la Proposición 2 de la página 200, si $X$ y $Y$ son **independientes**, entonces su correlación ($\rho$) y su covarianza son iguales a **0**.

---

### 2. Aplicación al Ejercicio de los Telescopios

En el ejercicio anterior, definimos:
*   **Telescopio Alfa ($X$):** $f_X(x) = 3x^2$ para $0 \le x \le 1$.
*   **Telescopio Beta ($Y$):** $f_Y(y) = 2y$ para $0 \le y \le 1$.
*   Se estableció que el tiempo de inicio de ambos instrumentos es **independiente** [Conversación previa].

#### Paso 1: Cálculo de las Medias ($\mu_X, \mu_Y$)
Calculamos el valor esperado de cada telescopio individualmente:
*   $E(X) = \mu_X = \int_{0}^{1} x(3x^2) \, dx = [ \frac{3x^4}{4} ]_0^1 = \mathbf{0.75}$
*   $E(Y) = \mu_Y = \int_{0}^{1} y(2y) \, dy = [ \frac{2y^3}{3} ]_0^1 = \frac{2}{3} \approx \mathbf{0.6667}$

#### Paso 2: Cálculo de $E(XY)$
Utilizando la función de densidad conjunta $f(x, y) = 6x^2y$ obtenida anteriormente:
$$E(XY) = \int_{0}^{1} \int_{0}^{1} (xy)(6x^2y) \, dy \, dx = \int_{0}^{1} \int_{0}^{1} 6x^3y^2 \, dy \, dx$$
Integrando primero respecto a $y$:
$$\int_{0}^{1} 6x^3y^2 \, dy = [ 2x^3y^3 ]_0^1 = 2x^3$$
Integrando ahora respecto a $x$:
$$\int_{0}^{1} 2x^3 \, dx = [ \frac{2x^4}{4} ]_0^1 = \mathbf{0.50}$$

#### Paso 3: Cálculo de la Covarianza
Aplicamos la fórmula abreviada:
$$Cov(X, Y) = E(XY) - \mu_X \cdot \mu_Y$$
$$Cov(X, Y) = 0.50 - (0.75) \cdot (\frac{2}{3}) = 0.50 - 0.50 = \mathbf{0}$$

#### Paso 4: Cálculo de la Correlación
Como la covarianza es cero, el numerador de la fórmula de correlación es cero:
$$\rho = \frac{0}{\sigma_X \cdot \sigma_Y} = \mathbf{0}$$

---

### 3. Interpretación de los Resultados

*   **Ausencia de Relación Lineal:** El hecho de que $\rho = 0$ indica que no existe ninguna relación lineal entre el tiempo de inicio del Telescopio Alfa y el del Telescopio Beta.
*   **Consistencia con la Independencia:** Este resultado matemático confirma la suposición inicial del ejercicio: dado que los telescopios operan de forma independiente, el conocimiento del tiempo de inicio de uno no proporciona ninguna información para predecir el tiempo del otro de manera lineal.
*   **Conclusión Científica:** En la planificación del observatorio, esto significa que los desfases o aciertos en el tiempo de captura de datos de un instrumento no influyen sistemáticamente en el rendimiento del otro.
