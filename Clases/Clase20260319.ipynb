### 1. Proposición de Valor Esperado para una Función $h(X, Y)$

La proposición que aparece en la **página 197** (dentro de la sección 5.2) establece las reglas para calcular el valor esperado de una función que depende de dos variables aleatorias conjuntamente distribuidas.

**Enunciado de la Proposición:**
Sean $X$ y $Y$ variables aleatorias conjuntamente distribuidas con función masa de probabilidad $p(x, y)$ (si son discretas) o función de densidad de probabilidad $f(x, y)$ (si son continuas). Entonces, el **valor esperado de una función $h(X, Y)$**, denotado por $E[h(X, Y)]$ o $\mu_{h(X, Y)}$, se calcula de la siguiente manera:

*   **Si $X$ y $Y$ son discretas:**
    $$E[h(X, Y)] = \sum_x \sum_y h(x, y) \cdot p(x, y)$$
*   **Si $X$ y $Y$ son continuas:**
    $$E[h(X, Y)] = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} h(x, y) \cdot f(x, y) \, dx \, dy$$

Esta proposición permite hallar el promedio a largo plazo de una función de dos variables sin necesidad de obtener primero la distribución de probabilidad de la propia función.

### 1. Definición de Covarianza (Página 198)
La **covarianza** entre dos variables aleatorias $X$ y $Y$ es:
$$Cov(X, Y) = E[(X - \mu_X)(Y - \mu_Y)]$$
Esta se calcula como:
*   **Caso discreto:** $\sum_x \sum_y (x - \mu_X)(y - \mu_Y)p(x, y)$
*   **Caso continuo:** $\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} (x - \mu_X)(y - \mu_Y)f(x, y) dx dy$.

### 2. Proposición: Fórmula abreviada de la Covarianza (Página 199)
La siguiente fórmula simplifica los cálculos al no requerir sustracciones intermedias:
$$Cov(X, Y) = E(XY) - \mu_X \cdot \mu_Y$$.

### 3. Definición de Correlación (Página 200)
El **coeficiente de correlación** de $X$ y $Y$, denotado por $Corr(X, Y)$, $\rho_{X,Y}$ o simplemente $\rho$, está definido por:
$$\rho_{X,Y} = \frac{Cov(X, Y)}{\sigma_X \cdot \sigma_Y}$$.

### 4. Proposiciones sobre la Correlación (Página 200)

**Proposición 1:**
1.  Si $a$ y $c$ son ambas positivas o ambas negativas, $Corr(aX + b, cY + d) = Corr(X, Y)$.
2.  Para dos variables aleatorias cualesquiera $X$ y $Y$, $-1 \le Corr(X, Y) \le 1$.

**Proposición 2:**
1.  Si $X$ y $Y$ son independientes, entonces $\rho = 0$, pero $\rho = 0$ no implica independencia.
2.  $\rho = 1$ o $-1$ si y sólo si $Y = aX + b$ para algunos números $a$ y $b$ con $a \neq 0$.

### 1. Proposición de Valor Esperado (Página 197)

La proposición fundamental para este cálculo establece que si $X$ y $Y$ son variables aleatorias continuas con una función de densidad de probabilidad (fdp) conjunta $f(x, y)$, el **valor esperado de cualquier función $h(X, Y)$** se calcula mediante la integral doble sobre todo el plano $xy$:

$$E[h(X, Y)] = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} h(x, y) \cdot f(x, y) \, dx \, dy$$

### 2. Adaptación al campo de la Astronomía

Dos telescopios robóticos, el **Telescopio Alfa** y el **Telescopio Beta**, están programados para iniciar una observación de una kilonova en el momento en que reciban una alerta de ondas gravitacionales. El tiempo de inicio de cada telescopio (medido como una fracción de la primera hora tras la alerta, $0 \le x, y \le 1$) es incierto y sigue distribuciones marginales distintas debido a sus diferentes sistemas de apuntado.

$X=$ Tiempo de inicio de la observación del telescopio alfa

$Y=$ Tiempo de inicio de la observación del telescopio beta

Supongamos que los tiempos de inicio son **independientes** y tienen las siguientes funciones de densidad:
*   $f_X(x) = 3x^2$ para $0 \le x \le 1$
*   $f_Y(y) = 2y$ para $0 \le y \le 1$

**Tarea:** Calcule el **tiempo esperado de desfase** entre el inicio de ambos telescopios. **Hint:** $h(X, Y) = |X - Y|$.

---

### 3. Resolución y Aplicación Matemática

#### Paso 1: Obtener la fdp conjunta
Como los tiempos de inicio son independientes, la función de densidad conjunta es el producto de las marginales:
$$f(x, y) = f_X(x) \cdot f_Y(y) = (3x^2)(2y) = 6x^2y$$
para $0 \le x \le 1$ y $0 \le y \le 1$.

#### Paso 2: Plantear la integral del valor esperado
Aplicando la proposición de la página 197:
$$E[|X - Y|] = \int_{0}^{1} \int_{0}^{1} |x - y| \cdot (6x^2y) \, dy \, dx$$

#### Paso 3: Resolver la integral dividiendo la región
Debido al valor absoluto $|x - y|$, debemos dividir la región de integración en dos partes: donde $x > y$ (desfase $x - y$) y donde $y > x$ (desfase $y - x$).

1.  **Región 1 ($y \le x$):**
    $$\int_{0}^{x} (x - y) 6x^2y \, dy = \int_{0}^{x} (6x^3y - 6x^2y^2) \, dy$$
    $$= [3x^3y^2 - 2x^2y^3]_0^x = 3x^5 - 2x^5 = \mathbf{x^5}$$

2.  **Región 2 ($y > x$):**
    $$\int_{x}^{1} (y - x) 6x^2y \, dy = \int_{x}^{1} (6x^2y^2 - 6x^3y) \, dy$$
    $$= [2x^2y^3 - 3x^3y^2]_x^1 = (2x^2 - 3x^3) - (2x^5 - 3x^5) = \mathbf{2x^2 - 3x^3 + x^5}$$

#### Paso 4: Integral externa respecto a $x$
Sumamos ambos resultados e integramos de 0 a 1:
$$E[|X - Y|] = \int_{0}^{1} (x^5 + 2x^2 - 3x^3 + x^5) \, dx = \int_{0}^{1} (2x^5 - 3x^3 + 2x^2) \, dx$$
$$E = \left[ \frac{2x^6}{6} - \frac{3x^4}{4} + \frac{2x^3}{3} \right]_0^1 = \frac{1}{3} - \frac{3}{4} + \frac{2}{3}$$
$$E = 1 - 0.75 = \mathbf{0.25}$$

### Interpretación Astronómica
El resultado de **0.25** indica que, bajo este modelo de probabilidad, el **desfase promedio esperado** entre el inicio de las observaciones de ambos telescopios es de **15 minutos** (0.25 de una hora). Esta información es crucial para la sincronización de datos multimensajero, permitiendo a los astrónomos prever cuánta información inicial del evento podría perderse o desfasarse entre los distintos instrumentos de la red global.

Para aplicar los conceptos de **covarianza** y **correlación** al ejercicio de los telescopios Alfa y Beta (basado en el modelo de variables continuas de nuestra conversación anterior), utilizaremos las definiciones y fórmulas fundamentales del libro de Devore.

### 1. Definiciones Teóricas (Fuente: Devore)

*   **Covarianza ($Cov(X, Y)$):** Es el valor esperado del producto de las desviaciones de las variables respecto a sus medias. Se utiliza la fórmula abreviada para simplificar los cálculos:
    $$Cov(X, Y) = E(XY) - \mu_X \cdot \mu_Y$$.
*   **Correlación ($\rho$):** Mide el grado de asociación lineal entre $X$ y $Y$. Se define como:
    $$\rho = \frac{Cov(X, Y)}{\sigma_X \cdot \sigma_Y}$$.
*   **Propiedad de Independencia:** Según la Proposición 2 de la página 200, si $X$ y $Y$ son **independientes**, entonces su correlación ($\rho$) y su covarianza son iguales a **0**.

---

### 2. Aplicación al Ejercicio de los Telescopios

En el ejercicio anterior, definimos:
*   **Telescopio Alfa ($X$):** $f_X(x) = 3x^2$ para $0 \le x \le 1$.
*   **Telescopio Beta ($Y$):** $f_Y(y) = 2y$ para $0 \le y \le 1$.
*   Se estableció que el tiempo de inicio de ambos instrumentos es **independiente** [Conversación previa].

#### Paso 1: Cálculo de las Medias ($\mu_X, \mu_Y$)
Calculamos el valor esperado de cada telescopio individualmente:
*   $E(X) = \mu_X = \int_{0}^{1} x(3x^2) \, dx = [ \frac{3x^4}{4} ]_0^1 = \mathbf{0.75}$
*   $E(Y) = \mu_Y = \int_{0}^{1} y(2y) \, dy = [ \frac{2y^3}{3} ]_0^1 = \frac{2}{3} \approx \mathbf{0.6667}$

#### Paso 2: Cálculo de $E(XY)$
Utilizando la función de densidad conjunta $f(x, y) = 6x^2y$ obtenida anteriormente:
$$E(XY) = \int_{0}^{1} \int_{0}^{1} (xy)(6x^2y) \, dy \, dx = \int_{0}^{1} \int_{0}^{1} 6x^3y^2 \, dy \, dx$$
Integrando primero respecto a $y$:
$$\int_{0}^{1} 6x^3y^2 \, dy = [ 2x^3y^3 ]_0^1 = 2x^3$$
Integrando ahora respecto a $x$:
$$\int_{0}^{1} 2x^3 \, dx = [ \frac{2x^4}{4} ]_0^1 = \mathbf{0.50}$$

#### Paso 3: Cálculo de la Covarianza
Aplicamos la fórmula abreviada:
$$Cov(X, Y) = E(XY) - \mu_X \cdot \mu_Y$$
$$Cov(X, Y) = 0.50 - (0.75) \cdot (\frac{2}{3}) = 0.50 - 0.50 = \mathbf{0}$$

#### Paso 4: Cálculo de la Correlación
Como la covarianza es cero, el numerador de la fórmula de correlación es cero:
$$\rho = \frac{0}{\sigma_X \cdot \sigma_Y} = \mathbf{0}$$

---

### 3. Interpretación de los Resultados

*   **Ausencia de Relación Lineal:** El hecho de que $\rho = 0$ indica que no existe ninguna relación lineal entre el tiempo de inicio del Telescopio Alfa y el del Telescopio Beta.
*   **Consistencia con la Independencia:** Este resultado matemático confirma la suposición inicial del ejercicio: dado que los telescopios operan de forma independiente, el conocimiento del tiempo de inicio de uno no proporciona ninguna información para predecir el tiempo del otro de manera lineal.
*   **Conclusión Científica:** En la planificación del observatorio, esto significa que los desfases o aciertos en el tiempo de captura de datos de un instrumento no influyen sistemáticamente en el rendimiento del otro.

La **covarianza** entre dos variables aleatorias $X$ y $Y$ se define en la página 198 del texto de Devore como el valor esperado del producto de las desviaciones de cada variable respecto a su media correspondiente.

La definición formal es la siguiente:

$$Cov(X, Y) = E[(X - \mu_X)(Y - \mu_Y)]$$

Su cálculo varía dependiendo de si las variables son discretas o continuas:

*   **Caso discreto:** Se obtiene mediante la suma doble $\sum_x \sum_y (x - \mu_X)(y - \mu_Y)p(x, y)$.
*   **Caso continuo:** Se obtiene mediante la integral doble $\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} (x - \mu_X)(y - \mu_Y)f(x, y) \, dx \, dy$.

Esta medida permite evaluar qué tan fuerte es la relación entre dos variables que no son independientes; un valor positivo indica que ambas tienden a variar en la misma dirección, mientras que un valor negativo sugiere una relación inversa.

La **covarianza** es una medida que evalúa la fuerza de la relación entre dos variables aleatorias $X$ y $Y$ cuando estas no son independientes. Matemáticamente, se define como el **valor esperado del producto de las desviaciones** de ambas variables respecto a sus medias correspondientes.

De acuerdo con la teoría de Devore, el signo del resultado permite interpretar la naturaleza de la relación entre las variables de la siguiente manera:

### 1. Covarianza Positiva ($Cov > 0$)
Ocurre cuando existe una **fuerte relación positiva**. Esto significa que los valores grandes de $X$ tienden a presentarse junto con valores grandes de $Y$, y los valores pequeños de $X$ con valores pequeños de $Y$. 
*   **En la fórmula:** Tanto $(x - \mu_X)$ como $(y - \mu_Y)$ tienden a ser ambos positivos o ambos negativos al mismo tiempo, resultando en un producto positivo.

### 2. Covarianza Negativa ($Cov < 0$)
Indica una **fuerte relación negativa** o inversa. En este caso, los valores grandes de una variable tienden a asociarse con valores pequeños de la otra.
*   **En la fórmula:** Los signos de las desviaciones tienden a ser opuestos (uno positivo y otro negativo), lo que genera un producto negativo.

### 3. Covarianza Cercana a Cero ($Cov \approx 0$)
Indica que las variables **no están fuertemente relacionadas de forma lineal**. Los productos positivos y negativos de las desviaciones tienden a eliminarse entre sí.
*   **Nota importante:** Si $X$ y $Y$ son **independientes**, su covarianza es exactamente **0**. Sin embargo, una covarianza de 0 **no implica necesariamente independencia**, ya que podría existir una relación no lineal muy fuerte que la covarianza no sea capaz de captar.

### Ejemplo con Astronomía: Masa y Luminosidad Estelar

Imagina que un astrónomo estudia una población de estrellas y define dos variables:
*   $X$: **Masa de la estrella** (en masas solares).
*   $Y$: **Luminosidad intrínseca** (energía emitida).

**Interpretación de los casos:**
*   **$Cov(X, Y) > 0$:** Este sería el escenario más común en la "Secuencia Principal". A medida que aumenta la masa de la estrella ($X$), también tiende a aumentar su luminosidad ($Y$). Existe una relación directa entre el tamaño/masa y el brillo.
*   **$Cov(X, Y) < 0$:** Supongamos que el astrónomo estudia una etapa exótica donde, por alguna razón física (como la pérdida de capas externas en una gigante roja), a medida que la masa central aumenta un poco por colapso, la luminosidad total detectada disminuye por oscurecimiento de polvo. Aquí la relación sería inversa.
*   **$Cov(X, Y) = 0$:** Si el astrónomo comparara la **Masa de la estrella** ($X$) con el **Número de lunas de un planeta** ($Y$) que orbita otra estrella en una galaxia distinta, no habría relación alguna. El conocimiento de la masa de una no ayuda a predecir la cantidad de lunas de la otra de forma lineal, por lo que serían **independientes** y su covarianza sería cero.

**Defecto de la Covarianza:** Un punto clave mencionado en las fuentes es que el valor numérico de la covarianza depende de las **unidades de medición** (por ejemplo, si mides la masa en kilogramos en lugar de masas solares, el valor de la $Cov$ cambiará drásticamente), por lo que para comparar la fuerza de la relación de forma estandarizada se utiliza la **correlación**.

La **definición del coeficiente de correlación** de $X$ y $Y$, denotado por $Corr(X, Y)$, $\rho_{X,Y}$ o simplemente $\rho$, aparece en la página 200 del texto y se establece de la siguiente manera:

$$\rho_{X,Y} = \frac{Cov(X, Y)}{\sigma_X \cdot \sigma_Y}$$

Esta medida representa la **covarianza "graduada a escala"** para que su valor no dependa de las unidades de medición originales de las variables, permitiendo así evaluar el **grado de asociación lineal** entre ellas.

Llegar a la fórmula de transformación para vectores de variables aleatorias es la extensión multidimensional de la regla para una sola variable que vimos anteriormente. Aunque el libro de Devore señala explícitamente que **no utiliza álgebra matricial** para mantenerse accesible a quienes solo tienen conocimientos de cálculo integral y diferencial, la lógica se fundamenta en el mismo principio de **cambio de variables en la integración** que se aplica en el caso univariante.

Para llegar a la expresión $g(y_1, \dots, y_n) = f(x_1, \dots, x_n) |\det(J)|$, se sigue este razonamiento basado en la teoría del cálculo y la probabilidad:

### 1. El principio de conservación de la probabilidad
En una variable aleatoria continua, la probabilidad de que $X$ caiga en un pequeño intervalo es el área bajo la curva de densidad. Para vectores, la probabilidad de que el vector $X$ caiga en una región $A$ del espacio $n$-dimensional debe ser igual a la probabilidad de que el vector transformado $Y$ caiga en la región correspondiente $B$.
$$\int \dots \int_A f(x_1, \dots, x_n) dx_1 \dots dx_n = \int \dots \int_B g(y_1, \dots, y_n) dy_1 \dots dy_n$$

### 2. Cambio de variables en cálculo multivariante
Al realizar un cambio de variables en una integral múltiple para pasar de las coordenadas $(x_1, \dots, x_n)$ a $(y_1, \dots, y_n)$, el elemento de volumen diferencial $dx_1 \dots dx_n$ no se traduce directamente como $dy_1 \dots dy_n$. Existe un factor de "estiramiento" o "compresión" del espacio.
Matemáticamente, la relación entre los diferenciales es:
$$dx_1 dx_2 \dots dx_n = \left| \det \left( \frac{\partial(x_1, \dots, x_n)}{\partial(y_1, \dots, y_n)} \right) \right| dy_1 dy_2 \dots dy_n$$

Donde el término de las derivadas parciales es la **Matriz Jacobiana ($J$)** de la transformación inversa ($X$ expresada en función de $Y$).

### 3. Analogía con el caso univariante
Podemos comparar esto directamente con la fórmula univariante que aparece en el **ejercicio 119 del Capítulo 4**:
*   **Univariante:** $g(y) = f[k(y)] \cdot |k'(y)|$, donde $k'(y)$ es la derivada de la función inversa $x=k(y)$.
*   **Multivariante:** El determinante del Jacobiano $|\det(J)|$ cumple exactamente la misma función que el valor absoluto de la derivada $|k'(y)|$. Es la medida de cuánto cambia el "volumen" de probabilidad al pasar de un sistema de coordenadas a otro.

### Resumen del proceso:
1.  **Definir la inversa:** Se tiene $Y = h(X)$, por lo que se debe hallar la función inversa $X = k(Y)$, es decir, expresar cada $x_i$ en términos de todas las $y_j$.
2.  **Calcular el Jacobiano:** Se construye la matriz de todas las derivadas parciales $\frac{\partial x_i}{\partial y_j}$ y se obtiene su determinante.
3.  **Sustituir:** La nueva densidad $g(y_1, \dots, y_n)$ será la densidad original evaluada en las funciones inversas, multiplicada por la magnitud de ese determinante para asegurar que la integral total siga siendo **1**.

Este experimento representa una aplicación práctica de la **transformación de variables aleatorias** analizada en la teoría de Devore. Para resolverlo, se siguen estos pasos basados en los principios de probabilidad presentados en las fuentes:

1.  **Definición de la variable original:** Al lanzar el lápiz "aleatoriamente", asumimos que el ángulo $\Theta$ que forma con el borde de la mesa sigue una **distribución uniforme**. Siguiendo el modelo del problema de la aguja de Buffon, el ángulo se distribuye en el intervalo $[0, \pi/2]$ con una función de densidad $f(\theta) = 2/\pi$.
2.  **La función de transformación:** La longitud de la proyección del lápiz sobre el borde se expresa como una función del ángulo: $Y = h(\Theta) = a \sin(\Theta)$, donde $a$ es la longitud constante del lápiz. 
3.  **Aplicación de la fórmula de transformación:** Como $h(\theta)$ es una función monótona e invertible en este intervalo, aplicamos la técnica para encontrar la nueva densidad $g(y)$ de la longitud proyectada.
    *   Se determina la función inversa: $k(y) = \arcsin(y/a)$.
    *   Se calcula la derivada de la inversa: $k'(y) = 1/\sqrt{a^2 - y^2}$.
4.  **Obtención de la nueva densidad:** La f.d.p. de la proyección $Y$ resulta de multiplicar la densidad original evaluada en la inversa por el valor absoluto de la derivada: $g(y) = (2/\pi) \cdot (1/\sqrt{a^2 - y^2})$ para $0 \le y \le a$.
5.  **Análisis estadístico:** Con esta nueva función $g(y)$, es posible calcular el **valor esperado** de la longitud de la proyección mediante la integral $\int y \cdot g(y) \, dy$, proporcionando el promedio de longitud que se observaría tras muchos lanzamientos.

Si el experimento considerara más variables aleatorias (como la posición del centro del lápiz respecto al borde), se aplicaría la **transformación multivariante** utilizando el **Jacobiano** para manejar la densidad conjunta de la posición y la proyección.

1.  **Simetría y Rango:** Aunque el ángulo $\Theta$ puede variar de $0$ a $2\pi$, la **longitud** de la proyección es una magnitud física que se repite en cada cuadrante. 2.  **Densidad de la variable original:** En el intervalo $(0, \pi/2)$, la función de densidad de probabilidad (f.d.p.) de un ángulo "aleatorio" es uniforme: $f(\theta) = \frac{1}{\pi/2 - 0} = \frac{2}{\pi}$.

5.  **Resultado final:** Multiplicando la densidad original por la derivada, obtenemos:

### Procedimiento paso a paso:

1.  **Definición de la integral:**
    Dado que estamos considerando el rango $0 < x < L$ y queremos hallar $P(X < L/2)$, los límites de integración son de $0$ a $L/2$:
    $$P(X < L/2) = \int_{0}^{L/2} \frac{2}{\pi \sqrt{L^2 - x^2}} \, dx$$

2.  **Extracción de constantes:**
    Sacamos el factor constante $2/\pi$ fuera de la integral:
    $$P(X < L/2) = \frac{2}{\pi} \int_{0}^{L/2} \frac{1}{\sqrt{L^2 - x^2}} \, dx$$

3.  **Resolución de la integral:**
    La antiderivada de $\frac{1}{\sqrt{L^2 - x^2}}$ es $\arcsin(x/L)$. Aplicamos el teorema fundamental del cálculo:
    $$P(X < L/2) = \frac{2}{\pi} \left[ \arcsin\left(\frac{x}{L}\right) \right]_{0}^{L/2}$$

4.  **Evaluación de los límites:**

    $$P(X < L/2) = \frac{2}{\pi} \left( \frac{\pi}{6} - 0 \right)$$

5.  **Resultado final:**

De acuerdo con la página 204 del libro de Devore (dentro de la Sección 5.3 "Estadísticos y sus distribuciones"), la definición de **estadístico** es la siguiente:

> "**Un estadístico** es cualquier cantidad cuyo valor puede ser calculado a partir de datos muestrales. Por consiguiente, un estadístico es una variable aleatoria y será denotada por una letra mayúscula; se utiliza una letra minúscula para representar el valor calculado u observado del estadístico".

De acuerdo con la definición proporcionada en la página 301 de las fuentes, un **estadístico** es cualquier cantidad cuyo valor puede ser calculado a partir de datos muestrales. Antes de obtener los datos, un estadístico se considera una variable aleatoria y se denota con una letra mayúscula.

A continuación se presentan 5 ejemplos de estadísticos mencionados en el texto:

1.  **Media muestral ($\bar{X}$)**: Es el promedio aritmético de las observaciones de una muestra y se utiliza frecuentemente como estimador de la media poblacional.
2.  **Desviación estándar muestral ($S$)**: Representa una medida de la dispersión de los datos dentro de la muestra en relación con la media muestral.
3.  **Varianza muestral ($S^2$)**: Se calcula a partir de la suma de las desviaciones al cuadrado de las observaciones respecto a la media, dividida por el tamaño de la muestra menos uno ($n-1$).
4.  **Rango muestral ($R$)**: Es un estadístico que mide la diferencia entre el valor máximo y el valor mínimo observados en el conjunto de datos de la muestra.
5.  **Proporción muestral ($X/n$ o $\hat{p}$)**: Es la fracción de individuos u objetos en una muestra que poseen una característica específica de interés.

De acuerdo con las fuentes proporcionadas, el enunciado formal del **Teorema del límite central (TLC)** aparece en un recuadro destacado (identificado en la secuencia del texto como parte de la sección 5.4) y establece lo siguiente:

**TEOREMA: Teorema del límite central (TLC)**

Sean $X_1, X_2, \dots, X_n$ una muestra aleatoria de una distribución con **media $\mu$** y **varianza $\sigma^2$**. Entonces, si $n$ es suficientemente grande:

*   **$\bar{X}$** tiene aproximadamente una **distribución normal** con $\mu_{\bar{X}} = \mu$ y $\sigma^2_{\bar{X}} = \sigma^2/n$.
*   **$T_o$** (el total de la muestra) también tiene aproximadamente una **distribución normal** con $\mu_{T_o} = n\mu$ y $\sigma^2_{T_o} = n\sigma^2$.

El teorema añade que **mientras más grande es el valor de $n$, mejor es la aproximación**. 

Este resultado es considerado el teorema de probabilidad más importante, ya que permite "pretender" que la media muestral es normal para calcular probabilidades incluso cuando la población original no es normal, proporcionando un "atajo impresionante" para la inferencia estadística.

## Ejemplo

Para explicar el **Teorema del Límite Central (TLC)** utilizando tu ejemplo de la edad de las estrellas en la Vía Láctea, imaginemos que eres un astrónomo tratando de entender la historia de nuestra galaxia.

### 1. La Población: Todas las estrellas de la Vía Láctea
Supongamos que la variable aleatoria $X$ representa la **edad de una estrella** individual. En la galaxia, la distribución de estas edades probablemente **no es normal** (no tiene forma de campana); podría estar sesgada porque hay muchísimas estrellas jóvenes en los brazos espirales y estrellas muy viejas en el halo o en cúmulos globulares. Esta población tiene una media verdadera $\mu$ (el promedio real de edad de todas las estrellas) y una desviación estándar $\sigma$.

### 2. El Experimento: Tomar muestras
Como es imposible medir cada una de las cientos de miles de millones de estrellas, tomas una **muestra aleatoria** de $n$ estrellas (por ejemplo, $n = 100$) y calculas su **edad promedio muestral** ($\bar{X}$).

### 3. ¿Qué dice el TLC en este contexto?
El teorema establece que, aunque la distribución de las edades de las estrellas sea "extraña" o asimétrica, si el tamaño de tu muestra $n$ es **suficientemente grande**, ocurrirá lo siguiente:

*   **Distribución Normal:** La edad promedio ($\bar{X}$) de tus 100 estrellas tenderá a seguir una **distribución normal** (una curva de campana), sin importar que la población original de estrellas no sea normal.
*   **Centro de la campana:** La media de todos los promedios posibles que podrías obtener será exactamente igual a la media real de la población $\mu$.
*   **Reducción de la dispersión:** La variabilidad de tu promedio ($\sigma^2_{\bar{X}}$) será mucho menor que la variabilidad de las estrellas individuales, calculándose como $\sigma^2/n$. Esto significa que, con una muestra grande, es muy probable que tu promedio esté muy cerca de la edad real de la galaxia.

### 4. ¿Cuándo es "suficiente" la muestra?
De acuerdo con la **regla empírica** de Devore, si la distribución de las edades de las estrellas es muy asimétrica, necesitarás que tu muestra sea de al menos **$n \ge 30$** para que la aproximación normal sea válida. Si las edades ya siguieran una distribución normal, el promedio sería normal incluso con muestras muy pequeñas.

### 5. ¿Para qué sirve esto al astrónomo?
El TLC es el "atajo impresionante" de la estadística. Gracias a él, puedes usar la **tabla normal estándar** para calcular la probabilidad de que tu estimación de la edad galáctica tenga un error determinado, permitiéndote hacer **inferencias precisas** sobre toda la Vía Láctea basándote solo en un pequeño grupo de estrellas.

En resumen: aunque las estrellas tengan edades caóticas y variadas, **los promedios de esas edades se comportan de manera predecible y ordenada bajo una curva normal**, siempre que observes suficientes estrellas a la vez.

```
(* Parámetros supuestos para la edad de las estrellas *)
mu = 8;        (* Media de la población *)
sigma = 3;     (* Desviación estándar de la población *)
n = 50;        (* Tamaño de la muestra *)

(* Parámetros de la distribución de la media muestral según el TLC [1] *)
muXbar = mu;
sigmaXbar = sigma / Sqrt[n];

(* Definición de la f.d.p. normal para la media muestral [5] *)
f[x_] := (1 / (sigmaXbar * Sqrt[2 * Pi])) * Exp[-(x - muXbar)^2 / (2 * sigmaXbar^2)];

(* Generación del gráfico (f, x_barra) *)
Plot[f[x], {x, muXbar - 4 * sigmaXbar, muXbar + 4 * sigmaXbar}, 
 PlotStyle -> {Thick, Darker@Orange}, 
 Filling -> Axis, 
 FillingStyle -> LightOrange,
 AxesLabel -> {"\!\(\*OverbarBox[\(x\)]\) (Edad promedio)", "f(\!\(\*OverbarBox[\(x\)]\))"}, 
 PlotLabel -> "Distribución de la Media Muestral - Edad de Estrellas", 
 LabelStyle -> {Bold, Medium},
 PlotRange -> All,
 ImageSize->600]
```
