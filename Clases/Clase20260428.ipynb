# Pruebas de hipótesis paramétricas con muestra simple

Cálculo de la probabilidad de cometer un **error de tipo II ($\beta$)** en pruebas de hipótesis para la media ($\mu$), específicamente para el **Caso 1** (población normal con desviación estándar $\sigma$ conocida).

### 1. Contexto y Estadístico de Prueba
Cuando se conoce la desviación estándar poblacional ($\sigma$), el estadístico de prueba utilizado para inferir sobre la media $\mu$ es el **estadístico $Z$**, el cual mide la distancia entre la media muestral ($\bar{x}$) y el valor nulo ($\mu_0$) en unidades de error estándar:

$$Z = \frac{\bar{x} - \mu_0}{\sigma/\sqrt{n}}$$

### 2. Definición de las Hipótesis
Para este análisis, utilizaremos una **prueba de cola superior**:
*   **Hipótesis Nula ($H_0$):** $\mu = \mu_0$
*   **Hipótesis Alternativa ($H_a$):** $\mu > \mu_0$

### 3. La Región de No Rechazo
En una prueba con nivel de significancia $\alpha$, la hipótesis nula **no se rechaza** si el valor calculado de $z$ es menor al valor crítico $z_\alpha$. 

Despejando la media muestral ($\bar{X}$), la condición para **no rechazar $H_0$** es:
$$\bar{X} < \mu_0 + z_\alpha \cdot \frac{\sigma}{\sqrt{n}}$$

### 4. Definición Analítica de $\beta$
La probabilidad de cometer un error de tipo II, denotada por $\beta$, ocurre cuando no se rechaza $H_0$ a pesar de que esta es falsa. Dado que $H_0$ es falsa, la verdadera media poblacional es un valor alternativo $\mu'$ (donde $\mu' > \mu_0$).

Por lo tanto, la probabilidad $\beta$ depende del valor específico de $\mu'$:
$$\beta(\mu') = P(\text{No Rechazar } H_0 \text{ cuando } \mu = \mu')$$
$$\beta(\mu') = P\left( \bar{X} < \mu_0 + z_\alpha \frac{\sigma}{\sqrt{n}} \bigg| \mu = \mu' \right)$$

### 5. Desarrollo de la Estandarización
Para calcular esta probabilidad, debemos estandarizar la variable $\bar{X}$ utilizando la **verdadera media $\mu'$**. Restamos $\mu'$ y dividimos por el error estándar en ambos lados de la desigualdad:

$$\beta(\mu') = P\left( \frac{\bar{X} - \mu'}{\sigma/\sqrt{n}} < \frac{\mu_0 + z_\alpha \frac{\sigma}{\sqrt{n}} - \mu'}{\sigma/\sqrt{n}} \right)$$

Simplificando el lado derecho:
$$\beta(\mu') = P\left( Z < z_\alpha + \frac{\mu_0 - \mu'}{\sigma/\sqrt{n}} \right)$$

### 6. Fórmula General y Cálculo Final
Basándose en la función de distribución acumulativa normal estándar ($\Phi$), la probabilidad de error tipo II para una prueba de cola superior se define como:

$$\beta(\mu') = \Phi\left( z_\alpha + \frac{\mu_0 - \mu'}{\sigma/\sqrt{n}} \right)$$

**Interpretación técnica:**
*   A medida que la verdadera media $\mu'$ se aleja del valor nulo $\mu_0$, el valor dentro de $\Phi$ se vuelve más negativo, lo que resulta en una **probabilidad $\beta$ menor** (es más fácil detectar el error).
*   Si el tamaño de la muestra ($n$) aumenta, el término del error estándar disminuye, lo que también reduce la probabilidad de cometer un error de tipo II.

Si $\mu \neq \mu'$ entonces $z$ no es normal estándar, luego $\frac{\overline{x}-\mu_0}{\sigma/\sqrt{n}}$ es estándar

Para el desarrollo del **Caso II**, que corresponde a pruebas de hipótesis sobre la media ($\mu$) cuando la desviación estándar poblacional ($\sigma$) es **desconocida** y el tamaño de la muestra es **grande**, se aplican los siguientes fundamentos analíticos según el texto de Devore:

### 1. Fundamento Teórico y Suposiciones
Cuando el tamaño de la muestra es grande ($n \ge 40$), el Teorema del Límite Central garantiza que la media muestral $\bar{X}$ tiene aproximadamente una distribución normal. Dado que $\sigma$ es desconocida, se utiliza la desviación estándar muestral ($S$) como su estimador. A pesar de que la sustitución de $S$ por $\sigma$ introduce variabilidad adicional, para muestras grandes esta diferencia es mínima, permitiendo que la variable estandarizada siga teniendo aproximadamente una distribución normal estándar.

### 2. Formulación de la Prueba
Para evaluar la hipótesis nula **$H_0: \mu = \mu_0$**, se utiliza el estadístico de prueba $Z$:

$$z = \frac{\bar{x} - \mu_0}{s/\sqrt{n}}$$

Donde:
*   **$\bar{x}$:** es la media calculada de la muestra.
*   **$\mu_0$:** es el valor nulo o de referencia.
*   **$s$:** es la desviación estándar de la muestra.
*   **$n$:** es el tamaño de la muestra ($n \ge 40$).

### 3. Regiones de Rechazo
El criterio de decisión se basa en el nivel de significancia ($\alpha$) y la hipótesis alternativa ($H_a$) planteada:

*   **Prueba de cola superior ($H_a: \mu > \mu_0$):** Se rechaza $H_0$ si $z \ge z_\alpha$.
*   **Prueba de cola inferior ($H_a: \mu < \mu_0$):** Se rechaza $H_0$ si $z \le -z_\alpha$.
*   **Prueba de dos colas ($H_a: \mu \neq \mu_0$):** Se rechaza $H_0$ si $z \ge z_{\alpha/2}$ o si $z \le -z_{\alpha/2}$.

### 4. Error de Tipo II ($\beta$) y Tamaño de Muestra
La determinación de la probabilidad de error tipo II ($\beta$) y el cálculo del tamaño de muestra necesario ($n$) se realizan utilizando las mismas fórmulas que en el caso de $\sigma$ conocida. En este procedimiento, se sustituye el valor de $\sigma$ por la desviación estándar muestral $s$ u otro valor plausible basado en conocimientos previos del proceso. El uso de $s$ es válido debido a que, con una muestra grande, este estimador es lo suficientemente preciso para no invalidar el nivel de confianza de la prueba.

Para el desarrollo del **Caso III**, que corresponde a pruebas de hipótesis sobre la media ($\mu$) cuando la desviación estándar poblacional ($\sigma$) es **desconocida** y el tamaño de la muestra es **pequeño** ($n < 40$), se requiere la suposición de que la población tiene una distribución normal.

### 1. Fundamento Teórico
Cuando la población es normal pero $\sigma$ es desconocida, no se puede utilizar el estadístico $Z$ basado en la curva normal estándar. En su lugar, se utiliza la desviación estándar muestral ($s$) para estimar $\sigma$. El estadístico resultante sigue una **distribución $t$ de Student** con $\nu = n - 1$ grados de libertad.

### 2. Formulación de la Prueba
Para evaluar la hipótesis nula **$H_0: \mu = \mu_0$**, se utiliza el estadístico de prueba **$t$**:

$$t = \frac{\bar{x} - \mu_0}{s/\sqrt{n}}$$

Donde:
*   **$\bar{x}$:** es la media calculada de la muestra.
*   **$\mu_0$:** es el valor nulo o de referencia.
*   **$s$:** es la desviación estándar de la muestra.
*   **$n$:** es el tamaño de la muestra.

### 3. Regiones de Rechazo
La decisión se basa en comparar el valor calculado de $t$ con los valores críticos de la **Tabla A.5** (Valores críticos para distribuciones $t$), utilizando $n-1$ grados de libertad:

*   **Prueba de cola superior ($H_a: \mu > \mu_0$):** Se rechaza $H_0$ si $t \ge t_{\alpha, n-1}$.
*   **Prueba de cola inferior ($H_a: \mu < \mu_0$):** Se rechaza $H_0$ si $t \le -t_{\alpha, n-1}$.
*   **Prueba de dos colas ($H_a: \mu \neq \mu_0$):** Se rechaza $H_0$ si $t \ge t_{\alpha/2, n-1}$ o si $t \le -t_{\alpha/2, n-1}$.

### 4. Determinación del Valor P
El **valor P** se obtiene como el área bajo la curva $t$ con $n-1$ grados de libertad en la dirección de la hipótesis alternativa. Debido a que las tablas de distribución $t$ suelen ser menos detalladas que las de la normal, a menudo se busca un rango para el valor P o se utiliza software estadístico para obtener el valor preciso.

### 5. Probabilidad de Error de Tipo II ($\beta$)
A diferencia de las pruebas $Z$, el cálculo de $\beta$ para la prueba $t$ es analíticamente complejo. Se recomienda el uso de las **curvas de potencia** (o gráficas de $\beta$) que se encuentran en la **Tabla A.17**, las cuales dependen del factor $d = |\mu_0 - \mu'|/\sigma$. Es necesario tener una estimación o valor plausible de $\sigma$ para poder utilizar estas gráficas.

Para ilustrar la aplicación del **Caso III** (inferencia sobre la media $\mu$ con $\sigma$ desconocida, muestra pequeña $n < 40$ y población normal), inventaremos un problema situado en el contexto de la evolución estelar, basándonos en la estructura de los ejercicios de las fuentes.

### Problema: La edad del Cúmulo Estelar "Andrómeda-X"

**Contexto:** Un equipo de astrofísicos teoriza que la edad promedio de las estrellas en un cúmulo estelar recientemente descubierto, denominado "Andrómeda-X", es de **300 millones de años**. Para validar esta teoría, se toma una muestra aleatoria de **10 estrellas** del cúmulo y se determinan sus edades mediante modelos de evolución estelar.

Debido a que el tiempo de observación es costoso, el tamaño de la muestra es pequeño ($n=10$) y la desviación estándar poblacional de las edades ($\sigma$) es desconocida. Se asume, por estudios previos en cúmulos similares, que las edades de las estrellas en esta región siguen una **distribución normal**.

**Datos de la muestra:**
*   **Tamaño de muestra ($n$):** 10 estrellas.
*   **Media muestral ($\bar{x}$):** 312 millones de años.
*   **Desviación estándar muestral ($s$):** 15 millones de años.
*   **Nivel de significancia ($\alpha$):** 0.05.

**Pregunta:** ¿Sugieren estos datos que la edad promedio verdadera del cúmulo difiere de los 300 millones de años propuestos?

---

### Resolución Analítica Paso a Paso

#### 1. Identificación del parámetro y el caso
El parámetro de interés es la edad promedio verdadera ($\mu$) de las estrellas en el cúmulo. Estamos ante el **Caso III** porque:
1.  La varianza poblacional ($\sigma^2$) es desconocida (usamos $s$).
2.  El tamaño de la muestra es pequeño ($n = 10 < 40$).
3.  Se supone una población normal.

#### 2. Formulación de hipótesis
*   **Hipótesis Nula ($H_0$):** $\mu = 300$ (La edad promedio es la propuesta por la teoría).
*   **Hipótesis Alternativa ($H_a$):** $\mu \neq 300$ (La edad promedio es diferente; prueba de dos colas).

#### 3. Estadístico de Prueba
Dado que no conocemos $\sigma$, utilizamos el estadístico **$t$ de Student**:
$$t = \frac{\bar{x} - \mu_0}{s/\sqrt{n}}$$
Donde $\mu_0 = 300$.

#### 4. Región de Rechazo y Valor Crítico
Para una prueba de dos colas con $\alpha = 0.05$ y **grados de libertad** $\nu = n - 1 = 9$:
*   Buscamos en la **Tabla A.5** (Valores críticos para distribuciones $t$) el valor $t_{\alpha/2, \nu} = t_{0.025, 9}$.
*   El valor crítico es **2.262**.
*   **Regla de decisión:** Se rechaza $H_0$ si $|t| \ge 2.262$.

#### 5. Cálculos
*   **Error estándar estimado:** $s / \sqrt{n} = 15 / \sqrt{10} \approx 15 / 3.162 \approx 4.743$.
*   **Cálculo de $t$:**
    $$t = \frac{312 - 300}{4.743} = \frac{12}{4.743} \approx \mathbf{2.53}$$

#### 6. Análisis del Valor P (Opcional pero recomendado)
Mirando la **Tabla A.8** para 9 grados de libertad:
*   Para $t = 2.53$, el área a la derecha está entre 0.01 y 0.025 (específicamente $P(T > 2.5) = 0.017$).
*   Como es una prueba de dos colas, el **Valor P** $\approx 2 \times 0.017 = \mathbf{0.034}$.

#### 7. Conclusión e Interpretación
*   **Decisión:** Como $t_{calc} = 2.53$ es mayor que el valor crítico $2.262$ (o bien, el Valor P $0.034 < 0.05$), **se rechaza la hipótesis nula $H_0$**.
*   **Interpretación en contexto:** Existe evidencia estadística suficiente al nivel del 5% para concluir que la edad promedio de las estrellas en el cúmulo "Andrómeda-X" es significativamente diferente de los 300 millones de años teorizados. Los datos sugieren que el cúmulo podría ser más viejo de lo que se pensaba originalmente.

En el problema inventado sobre la edad del cúmulo estelar "Andrómeda-X", **sí es posible visualizar e identificar plenamente cada componente** de la fórmula general del estadístico de prueba $Z$ (o en este caso específico, el estadístico $T$), ya que dicha expresión representa la estructura fundamental de cualquier prueba de hipótesis para la media.

Para visualizarla analíticamente, debemos mapear los elementos de la fórmula con los datos de nuestro problema astronómico de la siguiente manera:

1.  **El estimador puntual ($\hat{\theta}$):**
    *   En nuestro caso, corresponde a la **media muestral ($\bar{x}$)**, que es de **312 millones de años**. Es el valor que "calculamos" de las 10 estrellas observadas para representar a toda la población del cúmulo.

2.  **El valor nulo ($\theta_0$):**
    *   Corresponde al valor propuesto por la teoría de los astrofísicos, **$\mu_0 = 300$ millones de años**. Es el punto de referencia que queremos poner a prueba.

3.  **El error estándar ($\sigma_{\hat{\theta}}$):**
    *   Dado que no conocemos la desviación estándar poblacional ($\sigma$) y usamos la de la muestra ($s = 15$), visualizamos el **error estándar estimado $s_{\bar{x}} = s/\sqrt{n}$**.
    *   En el cálculo fue $15 / \sqrt{10} \approx \mathbf{4.743}$. Este valor actúa como nuestra "unidad de medida" de variabilidad.

### Visualización analítica del resultado:
La fórmula $Z = \frac{\hat{\theta} - \theta_0}{\sigma_{\hat{\theta}}}$ (que en el Caso III se denomina **$t$** para reflejar que usamos la distribución $t$ de Student) se visualiza así:

$$t = \frac{312 - 300}{4.743} = \frac{12}{4.743} \approx \mathbf{2.53}$$

*   **El numerador ($312 - 300 = 12$):** Es la distancia absoluta, en millones de años, entre lo que observamos y lo que la teoría predecía.
*   **La relación final ($2.53$):** Nos permite visualizar que la media observada está a **2.53 "unidades de error"** de distancia del valor teórico.

**Nota crítica de las fuentes:** Aunque la estructura es idéntica a la de $Z$, en este ejemplo visualizamos el resultado sobre una **curva $t$** (con 9 grados de libertad), la cual es un poco más plana y extendida en las colas que la curva $Z$ normal estándar, para compensar la incertidumbre adicional de haber estimado la varianza con una muestra pequeña.

Para realizar una prueba de hipótesis sobre la varianza poblacional ($\sigma^2$), se utiliza efectivamente la **distribución ji cuadrada ($\chi^2$)**, siempre que la muestra provenga de una **población con distribución normal**.

A continuación se detallan los elementos analíticos de este procedimiento:

### 1. Estadístico de Prueba
Cuando la población es normal, la variable aleatoria definida por la relación entre la varianza muestral ($S^2$) y la varianza poblacional ($\sigma^2$) sigue una distribución $\chi^2$ con $n - 1$ grados de libertad:

$$\chi^2 = \frac{(n-1)S^2}{\sigma_0^2}$$

Donde:
*   **$n$:** es el tamaño de la muestra.
*   **$S^2$:** es la varianza calculada de la muestra.
*   **$\sigma_0^2$:** es el valor nulo o de referencia propuesto en la hipótesis nula ($H_0$).

### 2. Formulación de la Prueba
*   **Hipótesis Nula ($H_0$):** $\sigma^2 = \sigma_0^2$.
*   **Hipótesis Alternativa ($H_a$):** Puede ser de cola superior ($\sigma^2 > \sigma_0^2$), cola inferior ($\sigma^2 < \sigma_0^2$) o de dos colas ($\sigma^2 \neq \sigma_0^2$).

### 3. Regiones de Rechazo
La decisión se basa en el nivel de significancia ($\alpha$) y los valores críticos de la **Tabla A.7**:

*   **Para $H_a: \sigma^2 > \sigma_0^2$:** Se rechaza $H_0$ si el $\chi^2$ calculado es $\ge \chi_{\alpha, n-1}^2$.
*   **Para $H_a: \sigma^2 < \sigma_0^2$:** Se rechaza $H_0$ si el $\chi^2$ calculado es $\le \chi_{1-\alpha, n-1}^2$.
*   **Para $H_a: \sigma^2 \neq \sigma_0^2$:** Se rechaza $H_0$ si el estadístico es $\ge \chi_{\alpha/2, n-1}^2$ o $\le \chi_{1-\alpha/2, n-1}^2$.

### Consideración Crítica
A diferencia de las pruebas para la media ($\mu$), las inferencias sobre la varianza son **altamente sensibles a la suposición de normalidad**. Si la población no es normal, los resultados de la prueba ji cuadrada pueden ser seriamente engañosos, incluso con muestras grandes.

A continuación se transcribe y resuelve el problema de manera analítica:

### Enunciado del Problema (Ejercicio 80, Cap. 8)
Para garantizar características razonablemente uniformes en una aplicación particular, se desea que la desviación estándar verdadera ($\sigma$) del punto de ablandamiento de un cierto tipo de alquitrán de petróleo sea **cuando mucho de 0.50°C**. Se determinaron los puntos de ablandamiento de **diez especímenes** independientes y se obtuvo una desviación estándar muestral (**$s$**) de **0.58°C**. ¿Contradice esto fuertemente la especificación de uniformidad? Pruebe las hipótesis apropiadas con un nivel de significancia de **$\alpha = 0.01$**.

---

### Resolución Analítica

**1. Identificación de parámetros e hipótesis:**
*   **Parámetro de interés:** Varianza poblacional $\sigma^2$ (o desviación estándar $\sigma$).
*   **Valor nulo:** $\sigma_0 = 0.50 \implies \sigma_0^2 = 0.25$.
*   **Hipótesis Nula ($H_0$):** $\sigma^2 = 0.25$ (La especificación de uniformidad se cumple).
*   **Hipótesis Alternativa ($H_a$):** $\sigma^2 > 0.25$ (La variabilidad es mayor a la permitida; prueba de cola superior).

**2. Estadístico de prueba:**
Dado que la población se asume normal, utilizamos el estadístico ji cuadrada:
$$\chi^2 = \frac{(n-1)s^2}{\sigma_0^2}$$

**3. Datos y cálculos:**
*   $n = 10 \implies$ grados de libertad $\nu = 9$.
*   $s = 0.58 \implies s^2 = 0.3364$.
*   $\sigma_0^2 = 0.25$.

Sustituyendo los valores:
$$\chi^2 = \frac{(10-1)(0.58)^2}{0.25} = \frac{9 \cdot 0.3364}{0.25} = \frac{3.0276}{0.25} = \mathbf{12.11}$$

**4. Región de rechazo:**
Para un nivel de significancia $\alpha = 0.01$ con $\nu = 9$ grados de libertad, buscamos el valor crítico $\chi_{0.01, 9}^2$ en la **Tabla A.7**:
*   $\chi_{0.01, 9}^2 = \mathbf{21.666}$.
*   **Regla de decisión:** Rechazar $H_0$ si el $\chi^2$ calculado $\ge 21.666$.

**5. Conclusión e Interpretación:**
Como el valor calculado (**12.11**) es **menor** que el valor crítico (**21.666**), **no se rechaza la hipótesis nula $H_0$**.

**Interpretación en contexto:** No existe evidencia estadística suficiente (al nivel del 1%) para afirmar que la desviación estándar del punto de ablandamiento excede los 0.50°C. Aunque la desviación observada en la muestra (0.58) es mayor al límite, la diferencia no es lo suficientemente grande como para descartar que se deba a la variabilidad natural del muestreo. La especificación de uniformidad no se considera contradicha.
