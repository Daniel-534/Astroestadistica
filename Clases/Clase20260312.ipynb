# Funciones de varias variables, distribuciones de probabilidad conjunta, distribuciones marginales. Combinación de variables aleatorias.

## Distribución de probabilidad conjunta (Discreto)

**Definición**

Sean $X$ y $Y$ dos variables aleatorias discretas definidas en el espacio muestral $S$ de un experimento. La **función masa de probabilidad conjunta** $p(x, y)$ se define para cada par de números $(x, y)$ como:

$$p(x, y) = P(X = x \text{ y } Y = y)$$

Debe cumplirse que $p(x, y) \ge 0$ y $\sum_x \sum_y p(x, y) = 1$.

Ahora sea $A$ cualquier conjunto compuesto de pares de valores $(x, y)$ (p. ej., $A = \{(x, y) : x + y = 5\}$ o $\{(x, y) : \text{máx}(x, y) \le 3\}$). Entonces la probabilidad $P[(X, Y) \in A]$ se obtiene sumando la función masa de probabilidad conjunta incluidos todos los pares en $A$:

$$P[(X, Y) \in A] = \sum \sum_{(x, y) \in A} p(x, y)$$

## Distribución marginal (Discreto)

**Definición**

Las **funciones masa de probabilidad marginal** de $X$ y de $Y$, denotadas por $p_X(x)$ y $p_Y(y)$, respectivamente, están dadas por:

$$p_X(x) = \sum_{y} p(x, y)$$
$$p_Y(y) = \sum_{x} p(x, y)$$

Para obtener la función masa de probabilidad marginal de $X$ evaluada en un valor particular $x$, se deben sumar las probabilidades conjuntas $p(x, y)$ sobre todos los valores posibles de $y$. De manera análoga, para obtener la de $Y$, se suma sobre todos los valores posibles de $x$.


## Función de densidad de probabilidad conjunta (Continuo)

Sean $X$ y $Y$ **variables aleatorias continuas**. Una **función de densidad de probabilidad conjunta** $f(x, y)$ de estas dos variables es una función que satisface $f(x, y) \ge 0$ y $\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f(x, y) \, dx \, dy = 1$. 

Para cualquier conjunto $A$ en dos dimensiones, la probabilidad de que el par $(X, Y)$ caiga en dicha región se obtiene mediante la integración de la función conjunta sobre el área:

$$P[(X, Y) \in A] = \iint_A f(x, y) \, dx \, dy$$

Si el conjunto $A$ es un rectángulo definido por los intervalos $[a, b]$ y $[c, d]$, la probabilidad se calcula como:

$$P(a \le X \le b, c \le Y \le d) = \int_{a}^{b} \int_{c}^{d} f(x, y) \, dy \, dx$$

## Funciones de densidad de probabilidad marginal (Continuo)

Las **funciones de densidad de probabilidad marginal** de $X$ y $Y$, denotadas por $f_X(x)$ y $f_Y(y)$ respectivamente, permiten conocer la distribución de una variable sin referencia a la otra. Estas se obtienen integrando la función de densidad conjunta con respecto a la variable que se desea excluir:

$$f_X(x) = \int_{-\infty}^{\infty} f(x, y) \, dy \quad \text{con } -\infty < x < \infty$$

$$f_Y(y) = \int_{-\infty}^{\infty} f(x, y) \, dx \quad \text{con } -\infty < y < \infty$$
