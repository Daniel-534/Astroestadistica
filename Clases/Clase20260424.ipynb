# Valor P

Es la probabilidad de encontrar un valor igual o más contradictorio que el medido suponiendo que $H_0$ es cierto

Depende de $H_a$.

Si $H_a$: $\theta\neq\theta_0$

$$P(\theta \geq \hat{\theta}^+ \text{ o } \theta \leq \hat{\theta}^-| H_0) = \text{ P-value}$$

Si el P-valor $< \alpha$ se rechaza $H_0$

El cálculo del **valor P** (o nivel de significación observado) depende de la naturaleza de la hipótesis alternativa ($H_a$) y de la distribución del estadístico de prueba utilizado ($Z$ para muestras grandes o $t$ para muestras pequeñas).

### Fórmulas para Pruebas $Z$
Si el estadístico de prueba tiene una distribución normal estándar ($Z$), el valor P se calcula según el tipo de prueba:

1.  **Prueba de cola superior ($H_a: \mu > \mu_0$):**
    $$P = 1 - \Phi(z)$$
    Es el área a la derecha del valor calculado $z$.
2.  **Prueba de cola inferior ($H_a: \mu < \mu_0$):**
    $$P = \Phi(z)$$
    Es el área a la izquierda del valor calculado $z$.
3.  **Prueba de dos colas ($H_a: \mu \neq \mu_0$):**
    $$P = 2[1 - \Phi(|z|)]$$
    Es el doble del área de la cola capturada por el valor absoluto de $z$.

*(Donde $\Phi(z)$ es la función de distribución acumulativa normal estándar)*.

### Fórmulas para Pruebas $t$
Para una prueba $t$ con $n-1$ grados de libertad, la lógica es idéntica pero utilizando las áreas bajo la curva de la **distribución $t$** en lugar de la normal:

1.  **Cola superior:** Área a la derecha de $t$ calculada.
2.  **Cola inferior:** Área a la izquierda de $t$ calculada.
3.  **Dos colas:** Suma de las áreas en ambas colas ($2 \times \text{área de una cola}$).

### Regla de Decisión
Una vez calculada la fórmula correspondiente, el criterio para rechazar la hipótesis nula ($H_0$) es comparar el valor P con el nivel de significación ($\alpha$) seleccionado:
*   Si **valor P $\le \alpha$**, se rechaza $H_0$.
*   Si **valor P $> \alpha$**, no se rechaza $H_0$.

$$\beta = P(\text{Tipo II}) = P(\text{No rechazar } H_0| H_0 \text{ Falso})$$
