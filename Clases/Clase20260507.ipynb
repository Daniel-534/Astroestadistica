Se hace el problema 6 del taller 4

La **distribución binomial negativa** se basa en un experimento estadístico que consiste en una secuencia de ensayos independientes con una probabilidad de éxito $p$ constante. A diferencia de la distribución binomial, donde el número de ensayos es fijo, en la binomial negativa el número de éxitos deseados ($r$) es fijo y el número de ensayos (o fallas) es aleatorio.

### Definición y Función masa de probabilidad
La variable aleatoria $X$ se define como el **número de fallas** que preceden al $r$-ésimo éxito. Su función masa de probabilidad (fmp) es:

$$nb(x; r, p) = \binom{x + r - 1}{r - 1} p^r (1 - p)^x \quad \text{para } x = 0, 1, 2, \dots$$.

### Propiedades Estadísticas
*   **Media ($E(X)$):** Representa el número esperado de fallas antes de alcanzar $r$ éxitos y se calcula como $\frac{r(1-p)}{p}$.
*   **Varianza ($V(X)$):** Mide la dispersión de las fallas y está dada por $\frac{r(1-p)}{p^2}$.
*   **Relación con la distribución geométrica:** En el caso especial donde $r = 1$ (esperar al primer éxito), la distribución binomial negativa se reduce a una **distribución geométrica**.

### Consideraciones adicionales
*   **Modelo generalizado:** La fmp está bien definida incluso cuando $r$ no es un entero, lo que permite ajustar datos en una amplia variedad de aplicaciones científicas e industriales.
*   **Estimación:** En la práctica, si no se conocen los parámetros, se pueden usar estimadores de momentos igualando la media y varianza poblacionales con las muestrales.

Para calcular paso a paso el valor esperado de la función $h(X) = \frac{r-1}{X+r-1}$, donde $X$ es una variable aleatoria con **distribución binomial negativa**, se utiliza la definición de valor esperado para variables discretas y las propiedades de los coeficientes binomiales.

A continuación se detalla el procedimiento analítico:

### 1. Definir la función masa de probabilidad (fmp)
La fmp de una variable binomial negativa $X$ (número de fallas antes del $r$-ésimo éxito) con parámetros $r$ y $p$ es:
$$p(x) = \binom{x + r - 1}{r - 1} p^r (1 - p)^x \quad \text{para } x = 0, 1, 2, \dots$$

### 2. Plantear la sumatoria del valor esperado
Según la definición de $E[h(X)]$, multiplicamos cada valor de la función por su probabilidad y sumamos sobre todo el dominio:
$$E\left[\frac{r-1}{X+r-1}\right] = \sum_{x=0}^{\infty} \left( \frac{r-1}{x+r-1} \right) \binom{x+r-1}{r-1} p^r (1-p)^x$$

### 3. Expandir el coeficiente binomial
Recordando que $\binom{n}{k} = \frac{n!}{k!(n-k)!}$, expandimos el término central:
$$\binom{x+r-1}{r-1} = \frac{(x+r-1)!}{(r-1)! x!}$$

Sustituyendo esto en la sumatoria:
$$E = \sum_{x=0}^{\infty} \frac{r-1}{x+r-1} \cdot \frac{(x+r-1)!}{(r-1)! x!} \cdot p^r (1-p)^x$$

### 4. Simplificación algebraica
Realizamos las siguientes cancelaciones:
*   $\frac{(x+r-1)!}{x+r-1} = (x+r-2)!$
*   $\frac{r-1}{(r-1)!} = \frac{1}{(r-2)!}$

La expresión se simplifica a:
$$E = \sum_{x=0}^{\infty} \frac{(x+r-2)!}{(r-2)! x!} p^r (1-p)^x$$

### 5. Identificar una nueva distribución binomial negativa
Observamos que $\frac{(x+r-2)!}{(r-2)! x!}$ equivale al coeficiente binomial $\binom{x+r-2}{r-2}$. Ahora factorizamos un valor de $p$ para ajustar la potencia:
$$E = p \cdot \sum_{x=0}^{\infty} \binom{x+(r-1)-1}{(r-1)-1} p^{r-1} (1-p)^x$$

### 6. Conclusión del cálculo
La sumatoria resultante corresponde a la suma total de las probabilidades de una distribución binomial negativa con parámetros $(r-1)$ y $p$. Dado que la suma de cualquier función masa de probabilidad sobre todo su rango es igual a **1**:
$$E = p \cdot (1) = p$$

**Resultado final:** El valor esperado $E\left[\frac{r-1}{X+r-1}\right]$ es igual a **$p$**. Esto demuestra que este estadístico es un **estimador insesgado** de la proporción poblacional $p$ en un modelo de distribución binomial negativa.

Para encontrar el **estimador de máxima verosimilitud (MLE)** de la proporción $p$ en una distribución binomial negativa, partimos de la función masa de probabilidad (fmp) definida en las fuentes:

### 1. Definición de la función de verosimilitud
La fmp de una variable aleatoria $X$ (número de fallas antes del $r$-ésimo éxito) es:
$$P(X=x) = \binom{x + r - 1}{r - 1} p^r (1 - p)^x$$

La **función de verosimilitud** $L(p)$ es el producto de las probabilidades de la muestra observada. Para una sola observación $x$, es proporcional a:
$$L(p) \propto p^r (1 - p)^x$$

### 2. Log-verosimilitud
Para facilitar el cálculo, aplicamos el logaritmo natural ($\ln$), lo que convierte el producto en una suma:
$$\ln L(p) = \ln \binom{x + r - 1}{r - 1} + r \ln(p) + x \ln(1 - p)$$

### 3. Derivación y maximización
Derivamos la función con respecto a $p$ e igualamos a cero para encontrar el máximo:
$$\frac{d}{dp} \ln L(p) = \frac{r}{p} - \frac{x}{1 - p} = 0$$

Resolviendo para $p$:
$$r(1 - p) = xp$$
$$r - rp = xp$$
$$r = p(r + x)$$
$$\hat{p}_{MLE} = \frac{r}{r + x}$$

### 4. Conclusión
El estimador de máxima verosimilitud para $p$ es:
$$\hat{p} = \frac{r}{n}$$
Donde **$n = r + x$** representa el **número total de ensayos** realizados hasta obtener los $r$ éxitos.

**Nota sobre el mensaje anterior:** Es importante distinguir que, aunque $\frac{r-1}{n-1}$ es un **estimador insesgado** de $p$ (como se demostró anteriormente), el **MLE** es simplemente la proporción observada de éxitos sobre el total de ensayos, $\frac{r}{n}$.

Para encontrar el **estimador de máxima verosimilitud (MLE)** de $p$ basado en una muestra aleatoria de $n$ observaciones $(x_1, x_2, \dots, x_n)$ de una distribución binomial negativa, el procedimiento es el siguiente:

### 1. Función de verosimilitud ($L$)
Como las observaciones son independientes, la probabilidad conjunta es el producto de las funciones de masa de probabilidad individuales,:
$$L(p) = \prod_{i=1}^n \binom{x_i + r - 1}{r - 1} p^r (1 - p)^{x_i}$$
Simplificando, agrupamos los términos con $p$ y $(1-p)$:
$$L(p) = \left[ \prod_{i=1}^n \binom{x_i + r - 1}{r - 1} \right] p^{nr} (1 - p)^{\sum x_i}$$

### 2. Log-verosimilitud ($\ln L$)
Aplicamos el logaritmo natural para convertir el producto en suma y facilitar la derivación,:
$$\ln L(p) = \ln(\text{constante}) + nr \ln(p) + \left( \sum x_i \right) \ln(1 - p)$$

### 3. Derivación y maximización
Derivamos respecto a $p$ e igualamos a cero para hallar el valor máximo,:
$$\frac{d}{dp} \ln L(p) = \frac{nr}{p} - \frac{\sum x_i}{1 - p} = 0$$

### 4. Solución para el estimador $\hat{p}$
Despejamos $p$:
$$nr(1 - p) = p \sum x_i$$
$$nr - nrp = p \sum x_i$$
$$nr = p(nr + \sum x_i)$$
$$\hat{p} = \frac{nr}{nr + \sum x_i}$$

### Conclusión
Si dividimos tanto el numerador como el denominador por $n$, obtenemos el estimador en términos de la media muestral de fallas ($\bar{x}$):
$$\hat{p}_{MLE} = \frac{r}{r + \bar{x}}$$

Este resultado es intuitivo: el estimador es la **proporción total de éxitos** ($nr$) sobre el **número total de ensayos realizados** en las $n$ muestras ($\sum x_i + nr$).
