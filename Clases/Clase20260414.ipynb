# Estimación por intervalos o intervalos de confianza

Para entender la variable aleatoria **media muestral ($\bar{X}$)**, es fundamental distinguirla de un simple promedio calculado. Mientras que $\bar{x}$ (minúscula) es un número fijo obtenido de una muestra específica, $\bar{X}$ (mayúscula) es una **variable aleatoria** porque su valor cambia de una muestra a otra antes de que los datos sean recolectados.

A continuación, se presenta un ejemplo basado en el comportamiento de un centro de servicio automotriz detallado en las fuentes para ilustrar este concepto.

### 1. El Escenario (La Población)
Imagina un taller que cobra tres precios fijos por una afinación, dependiendo del tipo de automóvil: **$40, $45 o $50**. No todos los clientes pagan lo mismo; la probabilidad de cada cobro es la siguiente:
*   20% pagan **$40** ($p=0.2$)
*   30% pagan **$45** ($p=0.3$)
*   50% pagan **$50** ($p=0.5$)

El promedio real de cobro en este taller (la media poblacional $\mu$) es **$46.50**.

### 2. El Experimento de Muestreo
Supongamos que seleccionamos al azar a **dos clientes** ($n=2$) que se atendieron hoy. Queremos saber cuál es el cobro promedio de esos dos clientes. Antes de elegirlos, no sabemos quiénes serán, por lo que el resultado de ese promedio es nuestra variable aleatoria $\bar{X}$.

### 3. Los posibles valores de $\bar{X}$
Como elegimos a dos personas, hay varias combinaciones posibles de lo que pudieron haber pagado. Cada combinación nos da un valor distinto para la media muestral $\bar{X}$:

*   **Caso A:** Ambos pagaron $40. Entonces $\bar{X} = (40+40)/2 = \mathbf{40.0}$.
*   **Caso B:** Uno pagó $40 y el otro $45. Entonces $\bar{X} = (40+45)/2 = \mathbf{42.5}$.
*   **Caso C:** Uno pagó $40 y el otro $50. Entonces $\bar{X} = (45.0)$.
*   **Caso D:** Ambos pagaron $50. Entonces $\bar{X} = \mathbf{50.0}$.

Existen otras combinaciones (como $45+$45 o $45+$50), lo que genera una lista de posibles valores para $\bar{X}$: **{40, 42.5, 45, 47.5, 50}**.

### 4. Por qué es una "Distribución de Muestreo"
Cada uno de estos promedios tiene su propia probabilidad de ocurrir. Por ejemplo, la probabilidad de que el promedio de los dos clientes sea exactamente $45 es del 29% ($0.29$), sumando las probabilidades de las combinaciones que dan ese resultado.

Si graficamos estos promedios y sus probabilidades, obtenemos la **distribución de muestreo de $\bar{X}$**. Al observar esta distribución, notarás propiedades asombrosas descritas por la estadística:

1.  **Centro:** Si promedias todos los promedios posibles de las muestras, el resultado será exactamente **$46.50**, igual que la media de la población original. Por esto se dice que $\bar{X}$ es un **estimador insesgado**.
2.  **Dispersión:** Los valores de los promedios están más "apretados" o concentrados que los precios originales. Mientras que los precios individuales varían entre 40 y 50, es mucho más difícil que un promedio de muchas personas se aleje del centro. Analíticamente, la varianza de $\bar{X}$ es la varianza original dividida por el tamaño de la muestra ($n$).
3.  **Forma:** Aunque la población original solo tiene tres valores (40, 45, 50) y no parece una "campana", la distribución de $\bar{X}$ empieza a verse más simétrica y acampanada. Según el **Teorema del Límite Central**, si aumentáramos el número de clientes en la muestra (por ejemplo a $n=30$), la distribución de $\bar{X}$ se vería casi exactamente como una **Distribución Normal**, sin importar que los precios originales no lo sean.

### Resumen conceptual
La variable aleatoria $\bar{X}$ representa el **proceso de promediar**. No es un número estático, sino una variable que fluctúa alrededor de la verdadera media poblacional ($\mu$), volviéndose más precisa y predecible (más parecida a una campana de Gauss) a medida que el tamaño de la muestra aumenta.

Supongamos una distribución normal para una R.V $X$, media y varianzaa $\mu$ y $\sigma^2$ 

$X\sim N(\mu,\sigma^2)$

La estimación por intervalos representa una técnica fundamental en la inferencia estadística que permite ir más allá de un simple valor numérico único. Una estimación puntual, por el hecho de ser un solo número, no proporciona información directa sobre la precisión y confiabilidad de dicha estimación. Debido a la variabilidad del muestreo, es virtualmente imposible que una estimación puntual coincida exactamente con el parámetro poblacional verdadero. Por esta razón, se prefiere calcular y reportar un intervalo completo de valores factibles conocido como intervalo de confianza (IC).

### 1. Concepto y Nivel de Confianza
Un intervalo de confianza se construye seleccionando primero un nivel de confianza, el cual cuantifica el grado de confiabilidad del procedimiento. Los niveles más frecuentemente utilizados en la práctica científica son 95%, 99% y 90%. Mientras más alto sea este nivel, mayor será la creencia de que el valor del parámetro se encuentra dentro del intervalo calculado. El ancho de este intervalo proporciona información crítica sobre la precisión: un intervalo angosto sugiere un conocimiento preciso del parámetro, mientras que uno muy amplio transmite una gran incertidumbre.

La interpretación técnica del nivel de confianza, por ejemplo del 95%, se basa en la frecuencia relativa a largo plazo. Esto significa que si el experimento de muestreo se repitiera una y otra vez, a la larga el 95% de los intervalos calculados contendrían el valor verdadero del parámetro poblacional. No se debe interpretar que el parámetro tiene una probabilidad del 95% de estar en un intervalo específico ya calculado, pues una vez que se sustituyen los valores muestrales, el intervalo es fijo y el parámetro es una constante desconocida.

### 2. Intervalos para la Media Poblacional ($\mu$)
La construcción de estos intervalos depende del tamaño de la muestra ($n$) y del conocimiento previo de la desviación estándar poblacional ($\sigma$).

*   **Muestra Grande con $\sigma$ Conocida:** Se fundamenta en la variable estandarizada $Z = (\bar{X} - \mu) / (\sigma / \sqrt{n})$, que sigue una distribución normal estándar. La expresión resultante para el intervalo es $\bar{x} \pm z_{\alpha/2} \cdot (\sigma / \sqrt{n})$.
*   **Muestra Grande con $\sigma$ Desconocida:** Cuando el tamaño de muestra es suficiente ($n \ge 40$), el teorema del límite central permite sustituir $\sigma$ por la desviación estándar muestral $s$ manteniendo la validez del procedimiento.
*   **Muestra Pequeña en Población Normal:** Si $n$ es pequeña y no se conoce $\sigma$, se utiliza la distribución $t$ con $n-1$ grados de libertad. El intervalo se expresa como $\bar{x} \pm t_{\alpha/2, n-1} \cdot (s / \sqrt{n})$. Las curvas de la distribución $t$ están más esparcidas que la normal estándar, lo que resulta en intervalos más anchos para compensar la incertidumbre adicional al estimar $\sigma$.

### 3. Intervalos para Proporciones Poblacionales ($p$)
Para estimar una proporción de "éxitos" en una población, se utiliza la proporción muestral $\hat{p} = X/n$. Si tanto $n\hat{p} \ge 10$ como $n(1-\hat{p}) \ge 10$, el estimador tiene una distribución aproximadamente normal. El intervalo tradicional es $\hat{p} \pm z_{\alpha/2} \sqrt{\hat{p}\hat{q}/n}$, aunque investigaciones recientes recomiendan fórmulas "mejoradas" (como la de Agresti-Coull) que ofrecen niveles de confianza reales más cercanos a los nominales, incluso en muestras pequeñas.

### 4. Intervalos para la Varianza ($\sigma^2$)
En el caso de una población normal, las inferencias sobre la varianza se basan en la distribución ji cuadrada ($\chi^2$) con $n-1$ grados de libertad. A diferencia de los intervalos para la media, estos no son simétricos debido a la asimetría de la distribución ji cuadrada. El intervalo resultante para $\sigma^2$ se define por los límites $[(n-1)s^2 / \chi^2_{\alpha/2, n-1}, (n-1)s^2 / \chi^2_{1-\alpha/2, n-1}]$.

### 5. Otros Tipos de Intervalos Estadísticos
Además de los intervalos de confianza, existen otras herramientas de estimación por intervalos:
*   **Intervalos de Predicción (IP):** Se utilizan para predecir el valor de una sola observación futura. Son considerablemente más anchos que los intervalos de confianza porque incluyen tanto la variabilidad de la estimación de la media como la variabilidad de la observación individual.
*   **Intervalos de Tolerancia:** Están diseñados para capturar un porcentaje específico (como el 90%) de todos los valores de la población con un nivel de confianza determinado.
*   **Intervalos Bootstrap:** Son técnicas intensivas en computación que permiten obtener intervalos de confianza sin necesidad de suponer una familia paramétrica (como la normal) para la distribución de la población.
*   **Intervalos No Paramétricos:** Basados en métodos como el de Wilcoxon, son válidos para una amplia gama de distribuciones siempre que sean continuas y simétricas.

De acuerdo con el **Teorema del Límite Central (TLC)**, la media muestral $\bar{X}$ sigue una distribución aproximadamente normal cuando el tamaño de la muestra $n$ es suficientemente grande, independientemente de la forma que tenga la distribución de la población original.

A continuación se detallan las características analíticas de esta distribución según las fuentes:

### 1. Forma de la distribución
Incluso si la población muestreada tiene una distribución altamente asimétrica o no normal, el cálculo de promedios produce una distribución más acampanada conforme $n$ aumenta. Para propósitos prácticos, se considera que la aproximación normal es válida si **$n \ge 30$** (regla empírica), aunque si la población está muy lejos de ser normal, se podría requerir una $n$ más grande (como $n > 40$ si se desconoce $\sigma$ y se usa $s$).

### 2. Parámetros de la distribución ($\bar{X}$)
Analíticamente, si la población original tiene una media $\mu$ y una varianza $\sigma^2$, la distribución de muestreo de $\bar{X}$ tiene los siguientes parámetros:
*   **Valor esperado (Media):** La distribución de $\bar{X}$ está centrada exactamente en la media de la población de la cual se seleccionó la muestra, es decir, $E(\bar{X}) = \mu_{\bar{X}} = \mu$.
*   **Varianza:** La variabilidad de la media muestral es menor que la de la población original y disminuye conforme aumenta el tamaño de la muestra: $V(\bar{X}) = \sigma^2_{\bar{X}} = \sigma^2/n$.
*   **Desviación estándar (Error estándar):** Se expresa como $\sigma_{\bar{X}} = \sigma / \sqrt{n}$.

### 3. Efecto del tamaño de la muestra ($n$)
El TLC establece que mientras más grande sea el valor de $n$, mejor será la aproximación a la curva normal. Esto implica dos fenómenos fundamentales observados en experimentos de simulación:
*   **Concentración:** $\bar{X}$ basado en una $n$ grande tiende a acercarse mucho más a $\mu$ que una $\bar{X}$ basada en una $n$ pequeña, debido al efecto de promediado de valores inusuales.
*   **Simetría progresiva:** Los histogramas de $\bar{X}$ se vuelven progresivamente menos asimétricos y más concentrados en torno al valor medio conforme $n$ se incrementa.

### 4. Estandarización
Gracias al TLC, para calcular probabilidades sobre $\bar{X}$ cuando $n$ es grande, se puede "pretender" que la distribución es normal y utilizar la variable estandarizada:
$$Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}$$
la cual tiene aproximadamente una distribución normal estándar $N(0, 1)$.

$$\bar{X} \sim N(\bar{x};\mu,\sigma^2/n)$$

La expresión $(\mu^- \leq \mu \leq \mu^+)$ representa un **intervalo de confianza (IC)** para la media poblacional $\mu$, donde $\mu^-$ corresponde al límite inferior y $\mu^+$ al límite superior de dicho intervalo.

A continuación, se detalla analíticamente el significado, la construcción y la interpretación de esta expresión según las fuentes:

### 1. Definición y Componentes
Un intervalo de confianza es un rango de valores factibles para un parámetro desconocido, calculado a partir de datos muestrales. En lugar de dar un solo número (estimación puntual), se reporta este intervalo para transmitir información sobre la **precisión** y la **confiabilidad** de la inferencia.

*   **$\mu^-$ (Límite inferior):** Es el valor mínimo del rango considerado plausible para $\mu$ con un nivel de confianza determinado.
*   **$\mu^+$ (Límite superior):** Es el valor máximo de dicho rango.
*   **Media poblacional ($\mu$):** Es el valor verdadero y constante (aunque desconocido) que se intenta capturar.

### 2. Construcción Analítica de los Límites
La forma de calcular $\mu^-$ y $\mu^+$ depende de la información disponible y del tamaño de la muestra:

*   **Población normal con desviación estándar ($\sigma$) conocida:**
    Los límites se definen como $\bar{x} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$, donde:
    $$\mu^- = \bar{x} - z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \quad y \quad \mu^+ = \bar{x} + z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \text{}$$
*   **Muestra grande ($n \ge 40$) con $\sigma$ desconocida:**
    Se utiliza la desviación estándar muestral $s$ y el teorema del límite central permite usar valores críticos de la distribución normal ($z$):
    $$\mu^- \approx \bar{x} - z_{\alpha/2} \cdot \frac{s}{\sqrt{n}} \quad y \quad \mu^+ \approx \bar{x} + z_{\alpha/2} \cdot \frac{s}{\sqrt{n}} \text{}$$
*   **Muestra pequeña ($n < 40$) en población normal con $\sigma$ desconocida:**
    Se debe utilizar la **distribución $t$ de Student** con $n-1$ grados de libertad:
    $$\mu^- = \bar{x} - t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}} \quad y \quad \mu^+ = \bar{x} + t_{\alpha/2, n-1} \cdot \frac{s}{\sqrt{n}} \text{}$$

### 3. Interpretación Técnica
La expresión indica que se tiene un nivel de confianza específico (comúnmente 90%, 95% o 99%) de que el parámetro $\mu$ se encuentra entre esos dos límites.

*   **Confiabilidad:** Un nivel de confianza del 95% significa que si se repitiera el proceso de muestreo infinitas veces y se calculara un intervalo para cada muestra, a la larga el 95% de esos intervalos contendría el valor verdadero de $\mu$.
*   **Precisión:** El **ancho del intervalo** $(\mu^+ - \mu^-)$ mide la precisión de la estimación. Un intervalo angosto indica un conocimiento preciso del parámetro, mientras que uno muy amplio refleja una gran incertidumbre.
*   **Estado del Parámetro:** Es importante notar que una vez calculados los límites con datos específicos, el intervalo es fijo. Por lo tanto, no se dice que "hay una probabilidad del 95% de que $\mu$ esté ahí", sino que se tiene un "95% de confianza", ya que $\mu$ no es una variable aleatoria, sino una constante desconocida.

La expresión de probabilidad $P(Z_c^- \leq Z \leq Z_c^+) = (1-\alpha)$ constituye el fundamento analítico para la construcción de **intervalos de confianza**, el método central de la estimación por intervalos. A continuación, se desarrolla este concepto detallando su origen, su manipulación matemática y su interpretación en términos de fiabilidad.

### 1. Definición de los Componentes
Para entender esta expresión, es necesario identificar qué representa cada término dentro del marco de la estadística inferencial:
*   **$Z$:** Es una variable aleatoria con distribución **normal estándar** ($N(0, 1)$). En el contexto de la media poblacional, surge de estandarizar la media muestral: $Z = \frac{\bar{X} - \mu}{\sigma/\sqrt{n}}$.
*   **$1-\alpha$:** Se denomina **nivel de confianza** (cuando se expresa como porcentaje, $100(1-\alpha)\%$) y cuantifica el grado de fiabilidad o "creencia" de que el intervalo capturará el parámetro verdadero.
*   **$Z_c^-$ y $Z_c^+$:** Son los **valores críticos** que capturan un área central de $(1-\alpha)$ bajo la curva normal. En la práctica estándar, debido a la simetría de la campana de Gauss, se definen como $Z_c^+ = z_{\alpha/2}$ y $Z_c^- = -z_{\alpha/2}$.

### 2. Desarrollo Analítico (Derivación del Intervalo)
El objetivo de la estadística es aislar el parámetro desconocido ($\mu$) a partir de la probabilidad conocida de la variable $Z$. El proceso se realiza mediante las siguientes manipulaciones algebraicas:

1.  **Planteamiento inicial:** Se parte de la probabilidad de que la variable estandarizada caiga entre los valores críticos:
    $$P\left(-z_{\alpha/2} \leq \frac{\bar{X} - \mu}{\sigma/\sqrt{n}} \leq z_{\alpha/2}\right) = 1-\alpha$$
2.  **Eliminación del denominador:** Se multiplica toda la desigualdad por el error estándar ($\sigma/\sqrt{n}$):
    $$P\left(-z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \leq \bar{X} - \mu \leq z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}\right) = 1-\alpha$$
3.  **Aislamiento de la media poblacional:** Se resta $\bar{X}$ en todos los términos:
    $$P\left(-\bar{X} - z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \leq -\mu \leq -\bar{X} + z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}\right) = 1-\alpha$$
4.  **Inversión de signos:** Al multiplicar por $-1$, la dirección de las desigualdades se invierte, resultando en la expresión final para el intervalo aleatorio:
    $$P\left(\bar{X} - z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \leq \mu \leq \bar{X} + z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}\right) = 1-\alpha$$

### 3. Significado del Nivel de Confianza $100(1-\alpha)\%$
La fiabilidad del intervalo no debe interpretarse como una probabilidad de que el parámetro esté en un rango ya calculado con datos específicos. Su significado técnico es el siguiente:
*   **Frecuencia relativa a largo plazo:** Si el experimento de muestreo se repitiera un número infinito de veces y se calculara un intervalo para cada muestra usando la misma fórmula, el $100(1-\alpha)\%$ de esos intervalos contendría el valor verdadero de $\mu$.
*   **Estado del intervalo:** Antes de recolectar los datos, el intervalo es **aleatorio** y tiene una probabilidad de $(1-\alpha)$ de cubrir a $\mu$. Una vez calculados los límites con valores observados ($\bar{x}$), el intervalo es **fijo** y se dice que se tiene un "nivel de confianza", ya que $\mu$ es una constante desconocida, no una variable aleatoria.

### 4. El Dilema entre Fiabilidad y Precisión
La expresión analítica muestra una relación directa entre el valor crítico y el ancho del intervalo:
*   **A mayor confianza, mayor ancho:** Si se aumenta el nivel de confianza (por ejemplo, de 95% a 99%), $\alpha$ disminuye, lo que hace que el valor crítico $z_{\alpha/2}$ aumente. Esto ensancha el intervalo, lo que incrementa la fiabilidad pero reduce la precisión de la estimación.
*   **Efecto del tamaño de muestra:** Para mantener una alta fiabilidad ($1-\alpha$) sin perder precisión (un intervalo angosto), la única vía es incrementar $n$, ya que esto reduce el error estándar $\sigma/\sqrt{n}$.

En resumen, la ecuación $P(Z_c^- \leq Z \leq Z_c^+) = (1-\alpha)$ es el mecanismo que permite transformar una probabilidad teórica sobre una distribución normal en una herramienta práctica para acotar el valor de un parámetro poblacional con un margen de error y una confianza controlados.

Para desarrollar el cálculo de $\alpha$ a partir de un nivel de confianza del 95% y explicar su significado basándose en las fuentes, seguiremos el procedimiento matemático y la interpretación técnica de la estadística inferencial.

### 1. Ejemplo práctico para despejar $\alpha$

El nivel de confianza se expresa comúnmente como un porcentaje dado por la fórmula $100(1-\alpha)\%$. Si deseamos trabajar con un nivel de confianza (P) del 95%, realizamos el siguiente despeje analítico:

1.  **Igualación de la fórmula al nivel deseado:**
    $$100(1-\alpha) = 95$$
2.  **Conversión a términos decimales (dividiendo por 100):**
    $$1-\alpha = 0.95$$
3.  **Despeje de $\alpha$:**
    $$\alpha = 1 - 0.95$$
    $$\alpha = 0.05$$

En el contexto de un intervalo de confianza bilateral, este valor de $\alpha$ se divide entre las dos colas de la distribución de probabilidad (usualmente la normal estándar $Z$ o la distribución $t$), de modo que cada cola contiene un área de $\alpha/2 = 0.025$. Para una distribución normal, este valor de $\alpha/2 = 0.025$ corresponde al valor crítico $z_{0.025} = 1.96$.

### 2. Significado conceptual de $\alpha$

Desde una perspectiva técnica y conceptual basada en el texto de Devore, $\alpha$ representa varios conceptos críticos en la inferencia:

*   **Probabilidad de error o "Falta de fiabilidad":** Si el nivel de confianza $(1-\alpha)$ mide el grado de fiabilidad de un intervalo, $\alpha$ es la probabilidad de que el procedimiento falle. Un $\alpha = 0.05$ significa que, en el largo plazo, el 5% de las muestras tomadas de la población producirán intervalos que **no** contienen el valor real del parámetro poblacional ($\mu$).
*   **Margen de riesgo:** $\alpha$ cuantifica el riesgo que el investigador está dispuesto a aceptar de cometer una inferencia incorrecta. En las pruebas de hipótesis, este mismo $\alpha$ se conoce como el **nivel de significación** y representa la probabilidad de cometer un **error de tipo I** (rechazar la hipótesis nula cuando es verdadera).
*   **Relación con la precisión:** Existe un compromiso entre $\alpha$ y el ancho del intervalo. Si intentamos reducir $\alpha$ (por ejemplo, de 0.05 a 0.01 para tener 99% de confianza), el valor crítico aumenta, lo que resulta en un intervalo más ancho y, por lo tanto, menos preciso.
*   **Interpretación de frecuencia relativa:** No se debe interpretar que el parámetro tiene una "probabilidad" de estar en un intervalo ya calculado; más bien, $\alpha$ indica que si el experimento de muestreo se repitiera un número infinito de veces, la proporción de intervalos "erróneos" tendería a ser exactamente $\alpha$.

En resumen, mientras que el nivel de confianza ($1-\alpha$) refleja nuestra seguridad en el procedimiento, $\alpha$ es la medida de la incertidumbre residual y el error potencial inherente al muestreo aleatorio.

Para mostrar analíticamente por qué $Z_c = 1.96$ cuando se trabaja con un nivel de confianza del 95%, debemos seguir el procedimiento de despeje de la probabilidad en una distribución normal estándar, basándonos en las definiciones y tablas proporcionadas en las fuentes.

### 1. Relación entre Nivel de Confianza y $\alpha$
El nivel de confianza se define como $100(1-\alpha)\%$. Para un nivel del 95%, realizamos el siguiente cálculo:
1.  $100(1-\alpha) = 95$
2.  $1-\alpha = 0.95$
3.  $\alpha = 1 - 0.95 = \mathbf{0.05}$

Este valor de $\alpha$ representa la probabilidad total de que el parámetro quede fuera del intervalo (el área de las colas).

### 2. Distribución del Error en las Colas
En un intervalo de confianza bilateral (de dos colas), el riesgo $\alpha$ se distribuye equitativamente en ambos extremos de la curva normal estándar ($Z$) para mantener la simetría.
*   Área en cada cola = $\alpha / 2$
*   Área en cada cola = $0.05 / 2 = \mathbf{0.025}$

### 3. Identificación del Valor Crítico ($Z_c$)
El valor crítico $Z_c$ (también denotado como $z_{\alpha/2}$) es el punto en el eje horizontal que deja un área de $0.025$ a su derecha. Analíticamente, esto significa que el área acumulada a la izquierda de este punto ($\Phi(z)$) debe ser:
*   $P(Z \leq Z_c) = 1 - 0.025 = \mathbf{0.975}$

Al consultar la **Tabla 4.1 "Percentiles normales estándar y valores críticos"** en las fuentes, se observa que para un área de cola ($\alpha$) de **0.025**, el valor crítico $z_{\alpha}$ correspondiente es exactamente **1.96**.

### 4. Verificación en la Distribución de Probabilidad
La fuente confirma esta relación al establecer la proposición fundamental para el intervalo de confianza:
$$P(-1.96 \leq Z \leq 1.96) = 0.95$$

Esto se debe a que:
*   El área entre $-1.96$ y $1.96$ bajo la curva normal estándar es exactamente $0.95$.
*   Si sumamos el área de la cola izquierda ($P(Z < -1.96) = 0.025$) y el área de la cola derecha ($P(Z > 1.96) = 0.025$), obtenemos el $\alpha = 0.05$ total.

**Conclusión:** Dado que un nivel de confianza del 95% implica que $\alpha/2 = 0.025$, el valor de $Z$ que captura esa probabilidad en las tablas de la distribución normal estándar es **1.96**. Este valor también se verifica en la tabla de distribuciones $t$ (Tabla A.5) en la fila de grados de libertad infinitos ($\infty$), donde bajo la columna de $0.025$ aparece el valor **1.960**.

Para transformar la expresión de probabilidad de modo que el parámetro $\mu$ quede en el centro de la desigualdad, se deben realizar una serie de manipulaciones algebraicas sobre la variable estandarizada $Z$ partiendo de la base de que $Z = \frac{\bar{X} - \mu}{\sigma/\sqrt{n}}$.

A continuación se detalla el desarrollo analítico según las fuentes:

1.  **Sustitución de Z:** Se reemplaza $Z$ por su definición en la desigualdad original:
    $$P\left(-1.96 \leq \frac{\bar{X} - \mu}{\sigma/\sqrt{n}} \leq 1.96\right) = 0.95$$

2.  **Eliminación del denominador:** Se multiplica toda la desigualdad por el error estándar ($\sigma/\sqrt{n}$):
    $$-1.96 \cdot \frac{\sigma}{\sqrt{n}} \leq \bar{X} - \mu \leq 1.96 \cdot \frac{\sigma}{\sqrt{n}}$$

3.  **Aislamiento de la media muestral:** Se resta $\bar{X}$ en todos los términos:
    $$-\bar{X} - 1.96 \cdot \frac{\sigma}{\sqrt{n}} \leq -\mu \leq -\bar{X} + 1.96 \cdot \frac{\sigma}{\sqrt{n}}$$

4.  **Inversión de signos:** Al multiplicar por $-1$ para obtener $\mu$ positiva, se invierte la dirección de las desigualdades:
    $$\bar{X} + 1.96 \cdot \frac{\sigma}{\sqrt{n}} \geq \mu \geq \bar{X} - 1.96 \cdot \frac{\sigma}{\sqrt{n}}$$

5.  **Expresión final:** Reordenando de menor a mayor, la probabilidad se escribe como:
    $$P\left(\bar{X} - 1.96 \cdot \frac{\sigma}{\sqrt{n}} \leq \mu \leq \bar{X} + 1.96 \cdot \frac{\sigma}{\sqrt{n}}\right) = 0.95$$

Esta expresión final representa el fundamento del **intervalo de confianza del 95%** para la media poblacional $\mu$. Indica que, antes de recolectar los datos, existe una probabilidad del 0.95 de que el intervalo aleatorio limitado por estos valores capture el valor verdadero de $\mu$. Una vez calculados los valores con una muestra específica ($\bar{x}$), el intervalo se vuelve fijo y se reporta como una estimación de intervalo.

4. I.C$(\mu) = \bar{X}\pm Z_{\alpha/2}\frac{\sigma}{\sqrt{n}}$
5. $Z_c = Z_{\alpha/2}$

El intervalo de confianza (IC) del $100(1-\alpha)\%$ para la media poblacional $\mu$ representa una de las herramientas fundamentales de la estadística inferencial para acotar el valor de un parámetro desconocido con un nivel de fiabilidad determinado. La construcción de este intervalo bajo las condiciones de una muestra grande y desviación estándar poblacional $\sigma$ conocida se detalla analíticamente a continuación:

### 1. El papel del Teorema del Límite Central (TLC)
Cuando el tamaño de la muestra $n$ es suficientemente grande (típicamente $n \ge 30$ o $n \ge 40$ según el rigor del autor), el TLC establece que la distribución de muestreo de la media $\bar{X}$ será aproximadamente normal. Esto es válido independientemente de si la población original tiene una distribución normal, asimétrica o de cualquier otra forma. Analíticamente, esto permite afirmar que la variable estandarizada:
$$Z = \frac{\bar{X} - \mu}{\sigma/\sqrt{n}}$$
sigue una distribución normal estándar $N(0, 1)$ con una media de 0 y una varianza de 1.

### 2. Derivación del Intervalo con $\sigma$ conocido
El supuesto de que $\sigma$ es conocido permite que el denominador de la variable estandarizada sea una constante fija, lo que simplifica la distribución de $Z$. Partiendo de la probabilidad de que $Z$ se encuentre entre dos valores críticos que capturan un área central de $1-\alpha$:
$$P(-z_{\alpha/2} \leq Z \leq z_{\alpha/2}) = 1 - \alpha \text{}$$

Al sustituir $Z$ por su definición y despejar el parámetro $\mu$, se obtienen los límites del intervalo aleatorio. Una vez que se recolectan los datos y se calcula el valor observado $\bar{x}$, el intervalo de confianza se expresa como:
$$IC(\mu) = \bar{x} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \text{}$$

### 3. Componentes de la Fórmula
*   **Media Muestral ($\bar{X}$):** Es el estimador puntual e insesgado de $\mu$ y constituye el centro del intervalo.
*   **Error Estándar ($\sigma/\sqrt{n}$):** Representa la desviación estándar de la media muestral y cuantifica la variabilidad del proceso de estimación.
*   **Valor Crítico ($Z_c$ o $z_{\alpha/2}$):** Es el valor en el eje horizontal de la curva normal estándar que deja un área de $\alpha/2$ en la cola superior. Por ejemplo, para un nivel de confianza del 95% ($\alpha = 0.05$), el valor crítico $z_{0.025}$ es exactamente 1.96.
*   **Margen de Error ($z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$):** También llamado límite en el error de estimación, define la distancia máxima a la que se espera que $\bar{x}$ se encuentre de $\mu$ con la confianza especificada.

### 4. Interpretación de la Fiabilidad y Precisión
El nivel de confianza $100(1-\alpha)\%$ no indica la probabilidad de que el parámetro esté en un intervalo ya calculado, sino la eficiencia del procedimiento a largo plazo. Si el experimento de muestreo se repitiera una y otra vez bajo las mismas condiciones, el $100(1-\alpha)\%$ de los intervalos resultantes capturaría el valor verdadero de la media poblacional $\mu$.

Existe un compromiso intrínseco entre la fiabilidad y la precisión: un aumento en el nivel de confianza (por ejemplo, del 95% al 99%) incrementa el valor crítico $z_{\alpha/2}$, lo que resulta en un intervalo más ancho y, por ende, menos preciso. Para mejorar la precisión (estrechar el intervalo) sin sacrificar la fiabilidad, la única vía analítica es incrementar el tamaño de la muestra $n$, lo que reduce el error estándar de forma proporcional a $1/\sqrt{n}$.

La estimación por intervalos en estadística busca capturar un parámetro poblacional desconocido, como la media ($\mu$), dentro de un rango de valores factibles calculado a partir de datos muestrales. En este proceso, la **fiabilidad** y la **precisión** son los dos pilares fundamentales que determinan la utilidad de la inferencia, y ambos están intrínsecamente vinculados a través de la estructura matemática del intervalo de confianza (IC).

### 1. Fiabilidad: El Nivel de Confianza
La fiabilidad de un intervalo de confianza se cuantifica mediante el **nivel de confianza**, seleccionado por el investigador antes de recolectar los datos.

*   **Definición y Niveles Comunes:** El nivel de confianza mide el grado de confiabilidad del procedimiento de estimación. Los niveles más utilizados en la práctica científica son 95%, 99% y 90%. Mientras más alto sea este porcentaje, mayor es la "creencia" o seguridad de que el valor real del parámetro se encuentra dentro de los límites calculados.
*   **Interpretación Técnica:** La fiabilidad no se refiere a la probabilidad de que un parámetro esté en un intervalo fijo ya calculado, sino a la eficiencia del método a largo plazo. Un nivel de confianza del 95% significa que si el experimento de muestreo se repitiera un número infinito de veces, el 95% de los intervalos resultantes capturaría el valor verdadero del parámetro.
*   **El Riesgo ($\alpha$):** Si un intervalo tiene un nivel de confianza del $100(1-\alpha)\%$, entonces $\alpha$ representa la probabilidad de que el intervalo calculado sea erróneo, es decir, que no contenga al parámetro. Por ejemplo, en un IC del 95%, existe un 5% de riesgo ($\alpha = 0.05$) de que la muestra seleccionada produzca un intervalo que falle en capturar a $\mu$.

### 2. Precisión: El Ancho del Intervalo
La precisión de la estimación se refleja directamente en el **ancho del intervalo** calculado ($w$).

*   **Significado del Ancho:** Un intervalo angosto sugiere que el conocimiento sobre el valor del parámetro es razonablemente preciso. Por el contrario, un intervalo muy amplio transmite una gran cantidad de incertidumbre, indicando que el estimador puntual puede estar muy alejado del valor real.
*   **Componentes del Ancho:** Analíticamente, para una media poblacional con desviación estándar ($\sigma$) conocida, el ancho se define como $w = 2 \cdot z_{\alpha/2} \cdot \sigma / \sqrt{n}$. De esta fórmula se derivan tres factores que afectan la precisión:
    1.  **Variabilidad de la población ($\sigma$):** A mayor dispersión en los datos, mayor será el ancho del intervalo y menor la precisión.
    2.  **Tamaño de la muestra ($n$):** Incrementar el número de observaciones es la forma más directa de mejorar la precisión, ya que el error estándar disminuye proporcionalmente a $1/\sqrt{n}$.
    3.  **Nivel de confianza ($z_{\alpha/2}$):** Como se detalla a continuación, aumentar la fiabilidad afecta negativamente la precisión.

### 3. El Dilema entre Fiabilidad y Precisión
Existe una relación inversa e inevitable entre estos dos conceptos: **la ganancia de confiabilidad acarrea una pérdida de precisión**.

*   **El costo de la certeza:** Si un investigador desea pasar de un 95% a un 99% de confianza para aumentar la fiabilidad, el valor crítico $z_{\alpha/2}$ aumenta (de 1.96 a 2.58). Esto resulta en un intervalo más ancho, lo que significa que la estimación es ahora menos precisa.
*   **El intervalo del 100%:** Técnicamente, el único intervalo con fiabilidad absoluta (100% de confianza) es $(-\infty, \infty)$. Aunque es totalmente fiable, carece por completo de precisión y utilidad informativa, ya que se sabía que el parámetro estaría allí incluso antes de tomar la muestra.
*   **Optimización del diseño:** Para maximizar la fiabilidad sin perder precisión (es decir, mantener un intervalo angosto), el estadístico debe recurrir a la determinación del tamaño de muestra necesario ($n$). Despejando $n$ de la fórmula del ancho, se puede calcular cuántas observaciones se requieren para alcanzar una precisión específica $w$ con un nivel de confianza determinado: $n = (2 \cdot z_{\alpha/2} \cdot \sigma / w)^2$.

En conclusión, la fiabilidad (nivel de confianza) garantiza la validez del procedimiento a largo plazo, mientras que la precisión (ancho) define qué tan útil es la información específica obtenida de la muestra actual para acotar el valor de un parámetro.

*   **Ganancia en Precisión:** Un intervalo más estrecho permite acotar el parámetro poblacional ($\mu$) en un rango de valores más pequeño, lo que estadísticamente se considera una estimación más precisa.
*   **Pérdida en Fiabilidad:** Al reducir el porcentaje (por ejemplo, al 90%), aumenta el riesgo ($\alpha$) de que el intervalo calculado no contenga el valor verdadero del parámetro. El "precio" de tener un intervalo más preciso es que tienes menos seguridad en que el procedimiento haya capturado realmente a $\mu$.

### 2. Si te refieres a un menor Ancho del Intervalo ($w$)
Un ancho menor es siempre el objetivo deseado, ya que indica que el conocimiento del valor del parámetro es razonablemente exacto. Analíticamente, un intervalo más pequeño o "menor" se puede obtener de tres formas:
*   **Disminuyendo la fiabilidad:** Como se explicó arriba, usando un valor crítico $z_{\alpha/2}$ más pequeño (lo cual es menos deseable).
*   **Incrementando el tamaño de la muestra ($n$):** Esta es la mejor forma de obtener un intervalo "menor". Al aumentar $n$, el error estándar ($\sigma/\sqrt{n}$) disminuye, lo que estrecha el intervalo sin tener que sacrificar el nivel de confianza.
*   **Poblaciones con menor variabilidad ($\sigma$):** Si los datos de la población están más concentrados, el intervalo será naturalmente más pequeño.

### Resumen del "Dilema del Estadístico"
Basado en las fuentes, existe una **relación inversa** entre la fiabilidad y la precisión:
*   **Intervalo "menor" (más estrecho):** Te da más **precisión**, pero si el tamaño de muestra es fijo, te da menos **fiabilidad**.
*   **Intervalo "mayor" (más ancho):** Te da más **fiabilidad** (más seguridad de capturar el parámetro), pero menos **precisión**.

En conclusión, un intervalo menor te da **más precisión informativa**, pero a menos que aumentes tu número de observaciones ($n$), esa precisión viene a costa de una **menor confianza** en el resultado.

# Intervalo de Confianza Unilateral

$s$ es la desviación estándar muestral

En el Capítulo 7 del libro de Devore, se establece que, aunque los intervalos de confianza bilaterales proporcionan un rango acotado por ambos lados para un parámetro, existen situaciones donde el interés del investigador se centra únicamente en una dirección de la estimación. A continuación, se presenta la explicación conceptual y el desarrollo analítico de los **intervalos de confianza unilaterales** (también conocidos como límites de confianza).

### 1. Explicación Conceptual
Un intervalo de confianza unilateral proporciona una cota o límite (ya sea superior o inferior) para el valor de un parámetro poblacional, como la media $\mu$, con un nivel de confianza determinado.

*   **Propósito:** Se utiliza cuando el objetivo del estudio es asegurar que un parámetro no exceda un valor máximo tolerable o que no sea inferior a un valor mínimo aceptable.
*   **Ejemplos de aplicación:** Un ingeniero de fiabilidad puede estar interesado únicamente en un **límite de confianza inferior** para la vida útil promedio de un componente (para garantizar que duren "al menos" tanto tiempo). Por el contrario, un psicólogo podría buscar un **límite de confianza superior** para el tiempo de reacción promedio a un estímulo (para asegurar que no sobrepasa cierto umbral).
*   **Diferencia con el bilateral:** Mientras que un intervalo bilateral distribuye el riesgo de error ($\alpha$) en ambas colas de la distribución, el intervalo unilateral concentra todo el riesgo $\alpha$ en una sola cola. Esto resulta en un límite que, para un mismo nivel de confianza, es más cercano a la media muestral que los límites de un intervalo de dos colas.

### 2. Desarrollo Analítico (Muestra Grande)
Para derivar estos límites en muestras grandes ($n \ge 40$), se utiliza la variable estandarizada $Z = \frac{\bar{X} - \mu}{S/\sqrt{n}}$, la cual sigue aproximadamente una distribución normal estándar según el Teorema del Límite Central.

#### A. Límite de Confianza Inferior
Se parte del hecho de que el área acumulada bajo la curva normal estándar a la izquierda del valor crítico $z_\alpha$ es de $1-\alpha$:
$$P(Z \le z_\alpha) = 1 - \alpha \text{}$$

1.  **Sustitución de $Z$:** $P\left( \frac{\bar{X} - \mu}{s/\sqrt{n}} \le z_\alpha \right) \approx 1 - \alpha$.
2.  **Despeje de $\mu$:** Multiplicando por el error estándar y restando $\bar{X}$, se llega a la desigualdad $\mu \ge \bar{x} - z_\alpha \cdot \frac{s}{\sqrt{n}}$.
3.  **Expresión final:** El intervalo es $(\bar{x} - z_\alpha \cdot \frac{s}{\sqrt{n}}, \infty)$.

#### B. Límite de Confianza Superior
De manera análoga, para una probabilidad de $1-\alpha$ de que la variable $Z$ sea mayor que $-z_\alpha$:
$$P(-z_\alpha \le Z) = 1 - \alpha \text{}$$

1.  **Manipulación:** Al despejar el parámetro, se obtiene la cota superior $\mu \le \bar{x} + z_\alpha \cdot \frac{s}{\sqrt{n}}$.
2.  **Expresión final:** El intervalo es $(-\infty, \bar{x} + z_\alpha \cdot \frac{s}{\sqrt{n}})$.

### 3. El Caso de Poblaciones Normales (Muestra Pequeña)
Si el tamaño de la muestra es pequeño pero se tiene evidencia de que la población es normal, el análisis se traslada a la **distribución $t$ de Student** con $n-1$ grados de libertad.

Las fórmulas correspondientes para un nivel de confianza de $100(1-\alpha)\%$ son:
*   **Límite superior:** $\mu \le \bar{x} + t_{\alpha, n-1} \cdot \frac{s}{\sqrt{n}}$.
*   **Límite inferior:** $\mu \ge \bar{x} - t_{\alpha, n-1} \cdot \frac{s}{\sqrt{n}}$.

### Resumen de Diferencias Clave
| Característica | Intervalo Bilateral | Intervalo Unilateral |
| :--- | :--- | :--- |
| **Área en la cola** | $\alpha/2$ en cada lado | $\alpha$ en un solo lado |
| **Valor Crítico** | $z_{\alpha/2}$ o $t_{\alpha/2, n-1}$ | $z_\alpha$ o $t_{\alpha, n-1}$ |
| **Precisión** | Acota el valor en un rango finito | Proporciona un rango infinito en una dirección |

Este enfoque unilateral permite al investigador ser más eficiente en el uso de los datos cuando el interés práctico solo requiere conocer una de las fronteras de la incertidumbre.

Con un nivel de confianza del **95%**, el valor máximo factible para la edad promedio verdadera ($\mu$) de todos los cúmulos estelares de la población es de **8.740 Gyr**. 

Esto significa que, aunque nuestra estimación puntual fue de 8.167, el procedimiento estadístico nos permite afirmar con una fiabilidad del 95% que la verdadera media no excede los 8.740 Gyr. En este tipo de intervalo unilateral, el límite inferior es técnicamente $-\infty$, ya que solo nos interesa acotar el error hacia valores superiores.

La expresión para el intervalo (o límite) de confianza unilateral surge analíticamente de la estandarización de la media muestral bajo condiciones donde la variabilidad poblacional es incierta. A continuación se detalla su origen, justificación y el significado de su notación.

### 1. Origen y Justificación Analítica
El fundamento de esta expresión es el estadístico **$T$**, definido como:
$$T = \frac{\bar{X} - \mu}{S/\sqrt{n}}$$
Donde $\bar{X}$ es la media muestral, $\mu$ la media poblacional, $S$ la desviación estándar muestral y $n$ el tamaño de la muestra.

**¿Por qué se usa esta expresión?**
*   **Incertidumbre en $\sigma$:** En la práctica, la desviación estándar poblacional ($\sigma$) es casi siempre desconocida. Al usar la desviación estándar de la muestra ($S$) para estimarla, se introduce una fuente adicional de variabilidad.
*   **Muestras pequeñas:** Cuando el tamaño de la muestra es pequeño ($n < 40$), no se puede invocar el Teorema del Límite Central para asegurar que el estadístico siga una distribución normal estándar ($Z$).
*   **Distribución t:** Bajo el supuesto de que la población de donde proviene la muestra es normal, el estadístico $T$ sigue lo que se denomina una **distribución $t$ de Student**. Esta distribución es más "esparcida" que la normal estándar para compensar la incertidumbre de estimar $\sigma$ con $S$.

### 2. Derivación del Límite Unilateral
Para un límite de confianza superior del $100(1-\alpha)\%$, partimos de la probabilidad de que la variable $T$ sea menor o igual a un valor crítico que captura un área de $1-\alpha$ a su izquierda:
$$P(T \leq t_{\alpha, n-1}) = 1 - \alpha$$

Al sustituir $T$ por su definición analítica y despejar el parámetro $\mu$, se obtiene la desigualdad:
$$\mu \le \bar{x} + t_{\alpha, n-1} \cdot \frac{s}{\sqrt{n}}$$

### 3. Explicación de la Notación $t_{\alpha, \nu}$
La notación utilizada en estas expresiones (como $t_{\alpha, n-1}$) identifica puntos específicos en el eje horizontal de la curva de densidad $t$ y tiene los siguientes componentes:

*   **La letra $t$:** Indica que el valor proviene de la familia de distribuciones $t$ de Student, las cuales tienen forma de campana y están centradas en cero, similar a la normal, pero con "colas" más gruesas.
*   **El subíndice $\alpha$ (Nivel de significancia):** Representa el área bajo la curva que queda en la **cola superior** (a la derecha) del valor crítico. Por ejemplo, en un límite unilateral del 95%, $\alpha = 0.05$. Todo el riesgo de "error" se concentra en un solo extremo de la distribución.
*   **El subíndice $\nu$ (o $n-1$):** Representa los **grados de libertad** (gl).
    *   Analíticamente, esto se refiere al número de desviaciones ($X_i - \bar{X}$) que pueden determinarse libremente.
    *   En este caso, se usa $n-1$ porque la desviación estándar $S$ se calcula usando la media muestral, lo que impone una restricción matemática a los datos.
    *   **Propiedad clave:** A medida que $\nu$ aumenta, la distribución $t$ se vuelve menos dispersa y se aproxima a la distribución normal estándar ($Z$). Por ello, para $\nu = \infty$, el valor de $t$ es idéntico al de $z$.
