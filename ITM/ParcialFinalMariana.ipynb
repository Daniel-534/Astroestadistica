```python
from scipy.stats import chi2

n = 16
s = 1.0
sigma0 = 6.0
alpha = 0.10

chi2_stat = ((n - 1) * s**2) / (sigma0**2)
chi2_crit = chi2.ppf(alpha, df=n-1)

print(f"Estadístico Chi-cuadrado: {chi2_stat:.4f}")
print(f"Valor crítico Chi-cuadrado: {chi2_crit:.4f}")
print(f"Chi2 < Chi2_crit: {chi2_stat < chi2_crit}")
print(f"\nDecisión: {'Se rechaza Ho (Falso)' if chi2_stat < chi2_crit else 'No se rechaza Ho (Verdadero)'}")
```

```python
import numpy as np

# Datos
evaluacion = np.array([1, 3, 4, 3, 3, 3, 2, 1, 1, 1, 1, 2])
tiempo = np.array([25, 15, 10, 10, 10, 9, 18, 20, 22, 24, 18, 17])

# Calcular coeficientes de regresión
n = len(evaluacion)
x_bar = np.mean(evaluacion)
y_bar = np.mean(tiempo)

b1 = np.sum((evaluacion - x_bar) * (tiempo - y_bar)) / np.sum((evaluacion - x_bar)**2)
b0 = y_bar - b1 * x_bar

# Predecir tiempo de respuesta cuando evaluación = 3.5
eval_pred = 3.5
tiempo_pred = b0 + b1 * eval_pred

print(f"b0 = {b0:.4f}")
print(f"b1 = {b1:.4f}")
print(f"Tiempo de respuesta para evaluación 3.5: {tiempo_pred:.4f}")
```

```python
import numpy as np
from scipy.stats import chi2

# Tabla de frecuencias observadas
observadas = np.array([[18, 15],
                       [25, 18]])

# Totales
total_filas = observadas.sum(axis=1)
total_columnas = observadas.sum(axis=0)
total_general = observadas.sum()

print(f"Total general: {total_general}")
print(f"Totales filas: {total_filas}")
print(f"Totales columnas: {total_columnas}")

# Calcular frecuencias esperadas
esperadas = np.outer(total_filas, total_columnas) / total_general
print(f"\nFrecuencias esperadas:")
print(esperadas)

# Calcular estadístico chi-cuadrado
chi2_stat = np.sum((observadas - esperadas)**2 / esperadas)
print(f"\nEstadístico chi-cuadrado: {chi2_stat:.4f}")

# Grados de libertad
filas, columnas = observadas.shape
gl = (filas - 1) * (columnas - 1)
print(f"Grados de libertad: {gl}")

# Valor crítico para α = 0.05
alpha = 0.05
chi2_critico = chi2.ppf(1 - alpha, gl)
print(f"Valor crítico (α=0.05): {chi2_critico:.4f}")

# Decisión
if chi2_stat > chi2_critico:
    print(f"\nDecisión: Rechazo Ho (χ²={chi2_stat:.4f} > {chi2_critico:.4f})")
else:
    print(f"\nDecisión: Acepto Ho (χ²={chi2_stat:.4f} < {chi2_critico:.4f})")
```
