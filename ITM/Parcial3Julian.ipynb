```python
import numpy as np
from scipy import stats

# Datos del problema
no_fumadoras = np.array([0.97, 1.16, 0.72, 0.86, 1.00, 0.85, 0.81, 0.58, 0.62, 0.57, 
                          1.32, 0.64, 0.99, 1.09, 0.90, 0.92, 0.74, 0.78, 0.88, 1.24])

fumadoras = np.array([0.48, 0.71, 0.98, 0.68, 1.18, 1.36, 0.78, 1.64, 0.94, 1.18, 1.24, 0.98])

# Estadísticas descriptivas
n1 = len(no_fumadoras)
n2 = len(fumadoras)
media1 = np.mean(no_fumadoras)
media2 = np.mean(fumadoras)
var1 = np.var(no_fumadoras, ddof=1)
var2 = np.var(fumadoras, ddof=1)

print("=== ESTADÍSTICAS DESCRIPTIVAS ===")
print(f"No fumadoras: n={n1}, media={media1:.4f}, varianza={var1:.4f}")
print(f"Fumadoras: n={n2}, media={media2:.4f}, varianza={var2:.4f}")

# Prueba t de Welch (varianzas desiguales)
# El estadístico t se calcula como:
t_stat = (media1 - media2) / np.sqrt(var1/n1 + var2/n2)

# Grados de libertad de Welch-Satterthwaite
df = ((var1/n1 + var2/n2)**2) / ((var1/n1)**2/(n1-1) + (var2/n2)**2/(n2-1))

# Valor-p para prueba de dos colas
p_valor = 2 * (1 - stats.t.cdf(abs(t_stat), df))

print("\n=== PRUEBA T DE WELCH (VARIANZAS DESIGUALES) ===")
print(f"Estadístico t: {t_stat:.4f}")
print(f"Grados de libertad (aprox): {df:.4f}")
print(f"Valor-p: {p_valor:.4f}")
print(f"Nivel de significancia: 0.01")

if p_valor > 0.01:
    print("\nDecisión: NO se rechaza H₀")
    print("Conclusión: No existen diferencias significativas")
else:
    print("\nDecisión: Se rechaza H₀")
    print("Conclusión: Existen diferencias significativas")
```
